### Principal Component Analysis - PCA
- This is most commonly used feature extraction technique to reduce the curse of dimensionality for feature extraction.
- This is a Unsupervised machine learning problem, uses complex mathemetics for feature extraction.
- PCA is a technique that reduces the data from higher dimensionality to lower dimensionality
- Advantages of using PCA:-
- - The number of columns reduces - Increases the faster execution of algorithms and less computation.
- - Visualization - reduces the features to 2D/3D which can be plotted and visualize.

### Feature selection VS Feature extraction
- <b>Feature selection</b> - Just select most important features that you think has significant impact on the outcome. But to prove this mathematically, we consider spread/variance.
- - Mathematically to prove the above
  - Take the spread/variance - the column that is spread more on the axis - which means that column is more say in deciding the outcome.
  - In below example, we take rooms column as it has more spread.
  - Feature selection is used when the spread of the features is different for different columns i.e. variance is different for different features, by just taking the feature that has more spread/variance..
  - 
- .
- <b>Feature extraction</b> - Algorithm reduces the features by creating new features. This is used when the spread of feature is almost same for all the features, which means there is a linear relationship between features, i.e. variance/spread is almost same for all features.
- In this case we cannot pick one column as both have the same variance/spread.
- 
- In this case we use feature extraction.

### How does PCA Work
- In above example, if we are unable to pick between number of rooms and number of washrooms, then derive a new column with the size of the flat by multiplying the size of rooms and washrooms, this creates a totally new column and then considers that in the model instead of taking no of rooms and number of washrooms features.
- PCA creates new features from the existing set of features and decides which new features have to be used in the model.

### Mathematical Intution / Geometric Intution behind PCA

- In the above example, PCA changes the axis, such that the variance of one feature is different from the variance of the other feature and then comes with new features.
- In this case, the No Of rooms axis is changed to rooms' and washrooms axis is changed to wrooms', because of this change, we can see the variance (d) of rooms' is greater than the variance (d') of wrooms', then these two new features are called as PCA1(rooms') and PCA2(wrooms'). PCA algorithm will choose PCA1(rooms') feature for the model building.

### CoVariance & CoVariance Matrix
- Covariance measures the direction of the relationship between two variables.
- A positive covariance means that both variables tend to be high or low at the same time.
- A negative covariance means that when one variable is high, the other tends to be low.
- CoVar(x,y) = Σ (X - μ of x)(Y - μ of y)/N
- +ve - both increase in the same direction.
- -ve - both in different directions.

### Eigen Decomposition of CoVariance Matrix
- When we apply matrices as a linear transformation, all the vectors (data points) are transformed to a new vectors.
- But, there will be 2 vectors in 2D space, which doesnt change the direction, these are called as Eigen vectors.
- Similarly, in 3D space, we get 3 eigen vectors, and so on.
- The corrosponding increased or decreased co-ordinates are called as Eigen values.
- Eigenvalue Equation:
- The eigenvalue equation is: Av = λv, where:
- A: is the square matrix.
- v: is the eigenvector (a non-zero vector that doesn't change direction when multiplied by A).
- λ: (lambda) is the eigenvalue.
### How Eigen vectors/values are used by PCA
- Use CoVariance Matrix of the dataset, instead of a normal random matrix for the Linear transformation, we get N number of Eigen vectors based on the dimension of the data.
- Pick the largest Eigen vector among all the eigen vectors, this eigen vector will always point to the direction of largest variance of the data. The magnitude of this Eigen vector will be the corrosponding Eigen value.

# Step wise derivation of Principal Components.
- Steps:-
- 1. Get the mean center of the data i.e. perform Standard Scaling of all the features.
  2. Derive CoVariance matrix.
  3. Get the Eigen Vectors & Values
  4. Choose those vectors/PC from the largest one as their variance will be more. Choosing PC 1 or 2 or more depends upon how much weightage each PC is giving us.
- Note:- The largest Eigen Vector will be the unit vector which has maximum spread/Variance. This is also called as Principal Component 1 (PC1)
- The second largest Eigen Vector will be called as PC2, and so on.
