```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
y
```

```python
# Scale the data

from sklearn.preprocessing import StandardScaler

StandardScaler_Obj = StandardScaler()
StandardScaler_Obj.fit(X)
X_Scaled = StandardScaler_Obj.transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
PCA_Obj = PCA(n_components=2)
```

```python
PCA_Obj.fit(X_Scaled)
```

```python
PCA_Obj.explained_variance_ratio_
```

```python
PCA_Obj.get_feature_names_out()
```

```python
PCA_Transformed = PCA_Obj.transform(X_Scaled)
```

```python
PCA_Transformed
```

```python
from sklearn.model_selection import train_test_split

x_train_NS, x_test_NS, y_train_NS, y_test_NS = train_test_split(X,y)
x_train_S, x_test_S, y_train_S, y_test_S = train_test_split(X_Scaled,y)
x_train_PCA, x_test_PCA, y_train_PCA, y_test_PCA = train_test_split(PCA_Transformed,y)
```

```python
from sklearn.linear_model import LogisticRegression

LogisticRegression_Obj_NS = LogisticRegression()
LogisticRegression_Obj_NS.fit(x_train_NS,y_train_NS)

LogisticRegression_Obj_S = LogisticRegression()
LogisticRegression_Obj_S.fit(x_train_S,y_train_S)

LogisticRegression_Obj_PCA = LogisticRegression()
LogisticRegression_Obj_PCA.fit(x_train_PCA,y_train_PCA)
```

```python
y_predict_NS = LogisticRegression_Obj_NS.predict(x_test_NS)

y_predict_S = LogisticRegression_Obj_S.predict(x_test_S)

y_predict_PCA = LogisticRegression_Obj_PCA.predict(x_test_PCA)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test_NS,y_predict_NS)
```

```python
accuracy_score(y_test_S,y_predict_S)
```

```python
accuracy_score(y_test_PCA,y_predict_PCA)
```
