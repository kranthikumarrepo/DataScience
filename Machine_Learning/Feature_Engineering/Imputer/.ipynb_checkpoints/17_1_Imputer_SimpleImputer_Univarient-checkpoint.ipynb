### Imputing
- Assign a value by following some process.
-
- <b>Handling missing/ null values
- 1. <b>Univarient (Single column):- The null value is replaced by mean/median/mode/const based on the values in the same column which has nulls.
- .
- <b>Use Simple Imputer Class
-    When data is Numeric > Apply Mean, Median and Mode to replace the null/ missing values.
-    Use Simple Imputer Class, when missing data is 5-10% and when MCAR
-    Advantages - Simple to use
-    Disadvantages - changes the distribution (shrinks the distribution as missing values are filled with mean/median which are at the center), which creates more outliers, impacts covariance/ CoRelation
- 
- <b>2. Multivarient (Multiple columns) - KNN
-    <b> Impute null values based on multiple columns & rows.
- .
- Complete case analysis
- Drop records with null values only when
- A. Missing values should be at random (MCAR - Missing Completely At Random)
- B. < 5% records have missing values

### Simple Imputer - Replace with Mean, Median, Mode(most_frequent) and Arbitrary(Constant) Value
### Impute using sklearn models, instead of dropna, fillna of Pandas.

```python
# Import Titanic DataSet

import pandas as pd
Titanic_DF = pd.read_csv(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\train.csv")
Titanic_DF.head(3)
```

```python
# Check if there are any null values in all the columns

Titanic_DF.isnull().sum()
```

```python
# Numeric - Age has null values, Impute once by Mean, then by Meadian and then by a constant.
# Categorical - Cabin and Embarked - no action here, keep them as it is.
# Drop all the Categorical columns and preapre the new data set with only numerical columns.

Titanic_DF_New = Titanic_DF.drop(columns=['Name','Sex','Ticket','Cabin','Embarked','PassengerId'])
Titanic_DF_New.head(6)
```

```python
# Train Test Split the data, create x = Pclass, Age, SibSp, Parch, Fare. and y = Survived

x = Titanic_DF_New.drop(columns=['Survived'],axis=1)
y = Titanic_DF_New['Survived']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
x_train.head(3)
```

```python
from sklearn.impute import SimpleImputer

# Imput with Mean
SimpleImputerObj_Mean = SimpleImputer(strategy='mean')
# Impute with Median
SimpleImputerObj_Median = SimpleImputer(strategy='median')
# Impute with Arbitary Value
SimpleImputerObj_Mode = SimpleImputer(strategy='most_frequent') # Can be applied on both categorical and numerical columns.
# Impute with Mode Value
SimpleImputerObj_Constant = SimpleImputer(strategy='constant',fill_value=99) # Use only when data is not missing at random
# df.Age.fillna(99)

SimpleImputerObj_Mean.fit(x_train)
SimpleImputerObj_Median.fit(x_train)
SimpleImputerObj_Constant.fit(x_train)
SimpleImputerObj_Mode.fit(x_train)
```

```python
SimpleImputerObj_Mean.statistics_
# Mean for col1,col2,....
```

```python
SimpleImputerObj_Median.statistics_
# Median for col1,col2,....
```

```python
SimpleImputerObj_Constant.statistics_
# Constant Value for col1, col2...
```

```python
SimpleImputerObj_Mode.statistics_
```

```python
x_train_transformed_Mean = SimpleImputerObj_Mean.transform(x_train)
x_test_transformed_Mean = SimpleImputerObj_Mean.transform(x_test)

x_train_transformed_Median = SimpleImputerObj_Median.transform(x_train)
x_test_transformed_Median = SimpleImputerObj_Median.transform(x_test)

x_train_transformed_Constant = SimpleImputerObj_Constant.transform(x_train)
x_test_transformed_Constant = SimpleImputerObj_Constant.transform(x_test)

x_train_transformed_Mode = SimpleImputerObj_Mode.transform(x_train)
x_test_transformed_Mode = SimpleImputerObj_Mode.transform(x_test)
```

```python
x_train.head(10)
```

```python
x_train_transformed_Mean_PD = pd.DataFrame(x_train_transformed_Mean, columns = x_train.columns)
x_train_transformed_Median_PD = pd.DataFrame(x_train_transformed_Median, columns = x_train.columns)
x_train_transformed_Constant_PD = pd.DataFrame(x_train_transformed_Constant, columns = x_train.columns)
x_train_transformed_Mode_PD = pd.DataFrame(x_train_transformed_Mode, columns = x_train.columns)
```

### Random imputation
- Random value from same column will be imputed in place of nulls.
- Can be applied to both numerical and categorical
- There is no sklearn class to implement this, we need to write python code.

#### Replace the null value records in Age column with a Random value - Using Pandas fillna

```python
#Titanic_DF_Age_NullsWithRandom[Titanic_DF_Age_NullsWithRandom.PassengerId.isin([5,6,7])]

# Import Titanic DataSet

import pandas as pd
Titanic_DF = pd.read_csv(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\train.csv")
Titanic_DF.head(3)
```

```python
# Create a new DF with only required columns

Titanic_DF_New = Titanic_DF.drop(columns=['Name','Sex','Ticket','Cabin','Embarked','PassengerId'],axis=1)
Titanic_DF_New.head(3)
```

```python
# Create a copy dataframe of x_train data

Titanic_DF_Copy = Titanic_DF_New.copy()
Titanic_DF_Copy.head(3)

# Titanic_DF_Copy and Titanic_DF are exactly same.
```

```python
# Create a duplicate column of Age on Titanic_DF_Copy with name as Age_Imputed

Titanic_DF_Copy['Age_Imputed'] = Titanic_DF_Copy.Age
Titanic_DF_Copy.head(3)
```

```python
# Get the total count of nulls in Age column

Titanic_DF_Copy.Age.isnull().sum()
```

```python
# Get the total count of nulls in Age_Imputed column

Titanic_DF_Copy.Age_Imputed.isnull().sum()
```

```python
# Select rows from Titanic_DF_Copy that has null values in Age_Imputed

Titanic_DF_Copy[Titanic_DF_Copy.Age_Imputed.isnull()].head(3)
```

```python
# Select only Age_Imputed column that has null values 

# Titanic_DF_Copy.Age_Imputed[Titanic_DF_Copy.Age_Imputed.isnull()].head(3)
# Or
Titanic_DF_Copy['Age_Imputed'][Titanic_DF_Copy.Age_Imputed.isnull()].head(3)
```

```python
# First remove the nulls from Age column.
# Since there are 177 nulls in Age_Imputed column, we need to get 177 random values from Age column
# Replace 177 nulls in Age_Imputed column with Randm 177 values from Age column

# LHS
# Titanic_DF_Copy.Age_Imputed[Titanic_DF_Copy.Age_Imputed.isnull()] >>> Selects recrods with 177 nulls in Age_imputed column with only Age_imputed column.
# RHS
# Titanic_DF['Age'].dropna() >>> Drop records with nulls in Age column - 177 records are dropped.
# Titanic_DF['Age'].dropna().sample() >>> After dropping 177 records, get sample from remaining records, but we should give sample(177) as we 
# need 177 samples, so get the count of nulls again within sample(code to get 177)
# Titanic_DF_Copy.Age.isnull().sum() >>> 177, put this code inside sample()
# Titanic_DF['Age'].dropna().sample(Titanic_DF_Copy.Age.isnull().sum()) >>> Gets 177 random values from Age, but along with index and other info.
# To get only 177 values, use .values
# Titanic_DF['Age'].dropna().sample(Titanic_DF_Copy.Age.isnull().sum()).values >>> Gets only 177 values.
# Assign above values to 177 nulls

Titanic_DF_Copy.Age_Imputed[Titanic_DF_Copy.Age_Imputed.isnull()] = Titanic_DF.Age.dropna().sample(Titanic_DF_Copy.Age.isnull().sum()).values
#Titanic_DF_Copy['Age_Imputed'][Titanic_DF_Copy['Age_Imputed'].isnull()] = Titanic_DF['Age'].dropna().sample(Titanic_DF_Copy['Age'].isnull().sum()).values
#Titanic_DF_Copy.loc[Titanic_DF_Copy['Age_Imputed'].isnull(),'Age_Imputed'] = Titanic_DF['Age'].dropna().sample(Titanic_DF_Copy['Age'].isnull().sum()).values
```
