#### KNN Imputer - For multivarient (Fill null based on multiple columns)
https://www.youtube.com/watch?v=-fK-xEev2I8
- <b>Iterative Imputer - One more Algorithm for Multivarient.
- Internally uses KNN Algorithm.
- 
- Above diagram shows NAN Euclidean distance, if we remove the weightage part(3/2,3/3, 3/1 and 3/2) it becomes normal distance.
- Calculates the Euclidean distance (Sqrt of (x2-x1)square+(y2-y1)square+.....) from the selected row containing null value(In above pic, 2nd row, feature1 has null) to distance of all the other rows.
- Example1:- if k = 2, it takes closest 2 rows(Least 2 distances) and takes mean of two values and updates here.
- If row1 distance from row2 = 11.29, row3 distance from row2 = 9, row4 distance from row2 = 5.19, row5 distance from row2 = 14.79. It takes row3 and row4 and then takes the value of Feature1 i.e. (23+40)/2 = 31.5 >>> This value will be kept in place of null in feature1 in row2.
- Example2:- if k = 1, it takes closest 1 row(Least distance) and takes the value and updates here.
- If row1 distance from row2 = 11.29, row3 distance from row2 = 9, row4 distance from row2 = 5.19, row5 distance from row2 = 14.79. It takes row4 and then takes the value of Feature1 i.e. 40 >>> This value will be kept in place of null in feature1 in row2.
- When there are nulls in neighbour rows, we calculate the <b>nan eculedian distance</b> Especially used for missing values, i.e. just gives weightage to the data points.
- dist(x,y) = sqrt(weight * sq. distance from present coordinates)
- weight = Total # of coordinates / # of present coordinates
- For example, the distance between [3, na, na, 6] and [1, na, 4, 5] is:

- <b>Advantages and Disadvantages:-
- KNN gives better results than mean, median and mode imputation.
- But takes more time as there are more calculations involved.
- Production, we need to keep the train data on the server, as it has to calculate the nearest neighbour every time it encounters a new null value.

```python
import pandas as pd
titanic_df = pd.read_csv(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\train.csv")[['Age','Pclass','Fare','Survived']]
titanic_df.head(3)
```

```python
titanic_df.isnull().sum()
# Only Age column has missing values
```

```python
x = titanic_df.drop(columns=['Survived'])
y = titanic_df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y)
x.head(6)
```

#### KNN Imputer along with weight option

```python
from sklearn.impute import KNNImputer

#KNNImputer_Obj = KNNImputer(n_neighbors=10) # Deafult is 5

# With weight option below - The neighbours that are closer get more priority/weightage while calculating the missing value.
# Example1:- if n_neighbors or k = 2, it takes closest 2 rows(Least 2 distances) and then gives weightage based on the distance, the one that is 
# more close get more weightage.
# If row1 distance from row2 = 11.29, row3 distance from row2 = 9, row4 distance from row2 = 5.19, row5 distance from row2 = 14.79. 
# It takes row3(dist=9) and row4(dist=5.19) and then takes the value of Feature1 
# i.e. (1/distance) * its value + (1/distance) * itsvalue....
# i.e. ((1/9)*23 + (1/5.19)*40)/2 = 2.55 + 7.7 = 10.257 >>> This value will be kept in place of null in feature1 in row2.
# If weight is not given, then (23+40)/2 = 31.5 >>> This value will be kept in place of null in feature1 in row2.
KNNImputer_Obj = KNNImputer(n_neighbors=1, weights='distance') # Deafult is 5

x_train_transform = KNNImputer_Obj.fit_transform(x_train)
x_test_transform = KNNImputer_Obj.fit_transform(x_test)
x_train_transform
```

```python
x_train_transform_DF = pd.DataFrame(x_train_transform,columns=x_train.columns)
x_train_transform_DF.head(6)
```

```python
# Check if there are any nulls after KNN Imputing

x_train_transform_DF.isnull().sum()
# Only Age column has missing values
```

```python
# Now since the nulls are imputed with KNN Nearest Neighbor, perform Logistic Regression to predict the values of y(Purchased)
```

```python
# Logistic Regression to predict values for y

from sklearn.linear_model import LogisticRegression
my_model = LogisticRegression()
my_model.fit(x_train_transform,y_train)
#my_model.fit(x_train_transform,y_train)
```

```python
y_predict = my_model.predict(x_test_transform)
y_predict
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_predict)
```

### If the nulls are imputed with SimpleImputer mean instead of KNN Imputer

```python
from sklearn.impute import SimpleImputer
SimpleImputer_Obj = SimpleImputer(strategy='mean')

x_train_trans_SI = SimpleImputer_Obj.fit_transform(x_train)
x_test_trans_SI = SimpleImputer_Obj.transform(x_test)
```

```python
my_model_si = LogisticRegression()
my_model_si.fit(x_train_trans_SI,y_train)
```

```python
y_predict_si = my_model_si.predict(x_test_trans_SI)
```

```python
accuracy_score(y_test,y_predict_si)
```
