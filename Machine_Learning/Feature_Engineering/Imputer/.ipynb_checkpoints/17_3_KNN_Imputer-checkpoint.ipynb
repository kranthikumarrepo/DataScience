#### KNN Imputer
- More notes in 17_1 and 17_2
- Internally uses KNN Algorithm
- Calculates the distance from the selected row, null value to distance to all rows, if k = 2, it takes closest 2 rows. and takes mean of two values and updates here.
- check todays youtube video
- better results than mean, median and mode imputation
- when there are nulls in neighbour rows, we calculate the nan eculedian distance
- dist(x,y) = sqrt(weight * sq. distance from present coordinates)
- weight = Total # of coordinates / # of present coordinates
- For example, the distance between [3, na, na, 6] and [1, na, 4, 5] is:



```python
import pandas as pd

Titanic_DF = pd.read_csv(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\train.csv")
Titanic_DF.head(9)
```

```python
x = Titanic_DF[['Age','Pclass','Fare']]
y = Titanic_DF['Survived']
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.impute import KNNImputer

KNNImputer_Obj = KNNImputer(n_neighbors=6, weights='distance')
```

```python
x_train_trns = KNNImputer_Obj.fit_transform(x_train)
x_test_trns = KNNImputer_Obj.transform(x_test)
x_test_trns
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, x_test_trns)
```
