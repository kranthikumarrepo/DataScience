### Feature Scaling - Normalization and Standardization 

https://www.youtube.com/watch?v=bqhQ2LWBheQ&list=PLTCVN6Wwg-pB7RXD5mYdP_Ve8S7z9GvxX&index=23
- Feature scaling is a data preprocessing technique used in machine learning to standardize the range of independent variables or features of a dataset.
- This ensures that all features contribute equally to the model training process, preventing features with larger scales from dominating the results.
- .
- Here's a more detailed explanation:
- .
- Why is feature scaling needed?
- .
- Prevent feature domination:
- Machine learning algorithms, especially those based on distance calculations (like k-means, KNN) or gradient-based optimization (like neural networks), can be heavily influenced by features with larger values. Feature scaling ensures that no single feature inappropriately biases the model's learning process.
- .
- Improve model performance:
- By bringing features to a similar scale, feature scaling can lead to better convergence and improved accuracy for many algorithms.
- .
- Avoid issues with distance calculations:
- In algorithms like k-means, Euclidean distance is used to measure similarity. If one feature has a much larger range than another, the distance calculation will be dominated by the feature with the larger range, potentially leading to incorrect clustering.
- .
- Common feature scaling methods:
- .
- <b>Normalization (Min-Max Scaling):</b>
- <b>Scales the data to a specific range, typically between 0 and 1.
- <b>This model is not good with outliers.</b>
- The formula is: X_scaled = (X - X_min) / (X_max - X_min).
- <b>Syntax:-<b/>
- from sklearn.preprocessing import MinMaxScaler
- scaler = MinMaxScaler()
- scaler.fit(data_df)
- data_df = scaler.transform(data_df)
- Or
- data_df = scalar.fit_transform(data_df) >>> Train and Transform in one syntax
- Scaled data (data_df) will be in the form of numpy array, convert it back to dataframe
- data_df = pd.DataFrame(data_df)
- .
- 
- .
- <b>Standardization - (Z-score normalization)</b>
- <b>Scales the data to have a mean of 0 and a standard deviation of 1. i.e. once the standardization is done, the mean of all the converted values will be 0 and SD = 1.
- <b> Not good results when the data has outliers. </b>
- The formula is: X_scaled = (X - mean) / standard_deviation.
- <b>Syntax</b>
- from sklearn.preprocessing import StandardScaler
- scaler = StandardScaler()
- scaler.fit(data_df)
- data_df = scaler.transform(data_df)
- Or
- data_df = scaler.fit_transform(data_df) >>> Train and Transform in one syntax
- Scaled data (data_df) will be in the form of numpy array, convert it back to dataframe
- data_df = pd.DataFrame(data_df)
- .
- 
- 
- Example:
- Consider a dataset with two features: "age" (ranging from 0 to 100) and "income" (ranging from 0 to 100,000). If you don't scale the data, the "income" feature will likely have a much larger impact on distance calculations or model parameters compared to the "age" feature. Scaling ensures that both features contribute equally to the model's learning process.
- .
- In essence, feature scaling is a crucial preprocessing step that ensures your machine learning models are trained on data that is fair and representative of all features, leading to more reliable and accurate results.
- .
- When to use Standardization and Normalization:-
- In case of neural networks, normalization is preferred since we dont assume any data distribution.
- Standardization is preferred when data follows Gaussian distribution(Normal distribution)
- Standardization is preferred over normalization when there are a lot of outliers.
- .
#### Other types of scaling
-
- <b>Robust Scaling:-</b>
- Robust scaling is used to handle outliers more effectively than standardization or Min-Max scaling.
- <b>Good with outliers but not good with sparse data(data with more zeros/nulls)</b>
- It uses the median and Interquartile Range (IQR) to scale the data, making it less sensitive to extreme values.
- The formula for robust scaling varies depending on the implementation, but it generally involves subtracting the median and dividing by the IQR.
- Formula:-
- <b>Xnew = (X - X Median)/IQR, Where IQR = Q3-Q1</b>
- .
- <b> MaxAbs Scaling:-</b>
- Scale each feature by its maximum absolute value. This estimator scales and translates each feature individually such that the maximal absolute value of each feature in the training set will be 1.0.
- <b> Good with sparse data(data with more zeros/nulls).
- <b> Not good with outliers
- <b> Scales data to the max of 1.
- Formula:-
- <b>x_new = xi / |xmax|</b>

### Example to show Normalization (Min Max Scaling)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

emp_data_df = pd.read_csv(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\ScalingData.csv")
emp_data_df.head(5)
# We can see all the data is at different scales.
```

```python
# Get the overview of the data

emp_data_df.describe()
```

```python
# Plot the pair plot

sns.pairplot(emp_data_df)
```

```python
# Normalize the data using the MinMaxScalar

from sklearn.preprocessing import MinMaxScaler

# Create object of MinMaxScalar class
MinMaxScalarObj = MinMaxScaler()

# fit and transform the emp_data_df dataframe and overwrite to the same emp_data_df dataframe
# Returns the scaled data, but in numpy.array format.
emp_data_df = MinMaxScalarObj.fit_transform(emp_data_df)

emp_data_df
```

```python
# Convert the numpy.array data format to pandas dataframe

#emp_data_df = pd.DataFrame(emp_data_df) #>>> If we do not give column names, it give default columns names starting from 0,1,...
emp_data_df = pd.DataFrame(emp_data_df, columns = ['SPPrice',	'Employment',	'InterestRates'])
```

```python
emp_data_df.head(3)
```

```python
# Get the overview of the data after scaling 

emp_data_df.describe()
```

```python
# Plot the pairplot after scaling

sns.pairplot(emp_data_df)
```
