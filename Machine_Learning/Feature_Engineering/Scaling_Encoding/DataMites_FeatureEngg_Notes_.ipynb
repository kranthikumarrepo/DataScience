# Normalization

### Normalization is a technique often applied as part of data preparation for machine learning.The goal of normalization is to change the value of numeric columns in the dataset to use a common scale, without distorting differences in the range of values or losing information

#### MinMaxScaling

MinmaxScaler is a normalization technique that scales the data to a fixed range typically [0,1]. It linearly
transforms features by shrinking them to fit within this range.

xi_new = xi - xmin / xmax - xmin

Sensitive to outliers

#### MaxAbs Scaling

MaxAbsScaler is a normalization technique that scales each feature by its maximun absolute value so that the resulting values are within the range [-1,1]

x_new = xi /|xmax|

#### Robust Scaling

RobustScaler is a normalization technique that scales features using stastics that are robust to outliers. it uses the median and interquartile range (IQR) instead of mean and standard deviation.


x_new = x - median / IQR
