### BoxPlot - Skewed Distribution
https://www.youtube.com/watch?v=Ccv1-W5ilak&t=684s
- Outliers when the data is skewed is calculated using Percentiles IQR Boxplot:-
- Lower Limit Outlier(Lower Whisker) = (Q1-1.5*IQR) and Upper Limit Outlier (Upper Whisker)= (Q3+1.5*IQR)
- Above formula is called IQR Proximity rule.
- In a box plot, any value above or below the whisker are called as outliers.
- Q1 = 25 Percentile value
- Q3 = 75 Percentile value
- IQR = Q3-Q1 = Value at Q3 - Value at Q1

- Ex: If value 98 percentile is 100, it means there are 100% values less than 98 value.
- Ex: If value 16 percentile is 0, it means there are no values less than 16

### Remove outliers using Trimming - Drop the records

```python
# First identify the outliers.
# To identify the outliers, we need to calculate Q1, Q3 and IQR

# Percentile25 = Placement_DF.describe().loc['25%']['placement_exam_marks'] 
# Or
PercentileQ1_25 = Placement_DF.placement_exam_marks.quantile(0.25)
print('PercentileQ1_25 is = ',PercentileQ1_25)
PercentileQ3_75 = Placement_DF.placement_exam_marks.quantile(0.75)
print('PercentileQ3_75 is = ',PercentileQ3_75)
IQR = PercentileQ3_75 - PercentileQ1_25
print('IQR is = ',IQR)

# Now calculate the Lower Limit Outlier(Lower Whisker) = (Q1-1.5IQR) and Upper Limit Outlier (Upper Whisker)= (Q3+1.5IQR)
LowerLimitOutlier = PercentileQ1_25 - 1.5*IQR
print('LowerLimitOutlier is ',LowerLimitOutlier)

UpperLimitOutlier = PercentileQ3_75 + 1.5*IQR
print('UpperLimitOutlier is ',UpperLimitOutlier)
```

```python
# Check the total outliers

# Placement_DF[Placement_DF.placement_exam_marks < LowerLimitOutlier ].shape # 0 records
Placement_DF[Placement_DF.placement_exam_marks > UpperLimitOutlier ].shape # 15 records
# There are 15 outliers in total.
```

##### Trim those records

```python
# Get only those recods where the placement_exam_marks is between -23.5 and 84.5

Placement_DF_TrimmedOutliers_IQR = Placement_DF[(Placement_DF.placement_exam_marks > LowerLimitOutlier) & (Placement_DF.placement_exam_marks < UpperLimitOutlier)]
Placement_DF_TrimmedOutliers_IQR.shape
```

#### Draw the distribution chart and box plot before dropping outlier data and after dropping the outlier data

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16,8))

# Plot distrbution plot and box plot for placement_exam_marks before trimming outliers
plt.subplot(2,2,1)
sns.histplot(Placement_DF.placement_exam_marks,kde=True)
plt.subplot(2,2,2)
sns.boxplot(x=Placement_DF.placement_exam_marks)

# Plot distrbution plot and box plot for placement_exam_marks after trimming outliers
plt.subplot(2,2,3)
sns.histplot(Placement_DF_TrimmedOutliers_IQR.placement_exam_marks,kde=True)
plt.subplot(2,2,4)
sns.boxplot(x=Placement_DF_TrimmedOutliers_IQR.placement_exam_marks)
```

```python
# We can see values > 84.5 are not present in the histogram and we can see outliers not present in the boxplot.
```

### Cap the outliers using Min and Max Value - Do not drop but replace the outliers with min and max values

```python
# Replace the outliers using the max and min values

# First find the outlier records
Placement_DF[(Placement_DF.placement_exam_marks < LowerLimitOutlier) | (Placement_DF.placement_exam_marks > UpperLimitOutlier)].shape
```

```python
# Create a copy of Placement_DF and add Capped_placement_exam_marks column

Placement_DF_New = Placement_DF.copy()
Placement_DF_New['Capped_placement_exam_marks'] = Placement_DF.placement_exam_marks
```

```python
# Now replace the lower and higher values on outliers.

# np.where(condition, If_True, IF_False)
# np.where(a < 5, a, 10*a)

import numpy as np
Placement_DF_New['Capped_placement_exam_marks'] = np.where(Placement_DF.placement_exam_marks < LowerLimitOutlier, LowerLimitOutlier, 
 np.where(Placement_DF.placement_exam_marks > UpperLimitOutlier, UpperLimitOutlier, Placement_DF.placement_exam_marks))
```

```python
# Check upper limit values are replced
Placement_DF_New[Placement_DF_New.placement_exam_marks > 84.5]
```

#### Draw the distribution chart and box plot before capping outlier data and after capping the outlier data

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16,8))

# Plot distrbution plot and box plot for placement_exam_marks before capping outliers
plt.subplot(2,2,1)
sns.histplot(Placement_DF_New.placement_exam_marks,kde=True)
plt.subplot(2,2,2)
sns.boxplot(x=Placement_DF_New.placement_exam_marks)

# Plot distrbution plot and box plot for placement_exam_marks after capping outliers
plt.subplot(2,2,3)
sns.histplot(Placement_DF_New.Capped_placement_exam_marks,kde=True)
plt.subplot(2,2,4)
sns.boxplot(x=Placement_DF_New.Capped_placement_exam_marks)
```

```python
# We can see more values around 84.5 in the histogram and from the box plot we can see all the outliers are removed.
```
