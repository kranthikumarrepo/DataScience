- <b> How to detect Outliers?
- 1. When the distribution is Normal Distribution:- ND Plot, Z Score - Any value beyond (Mean + 3SD) Or (Mean - 3SD) is an outlier.
- 
- 2. When the distribution is Skewed Distribution - BoxPlot, Any value beyond whisker (Q1-1.5\*IQR) Or (Q3+1.5\*IQR) is an outlier. Where IQR = Q3-Q1
     
  3. For Other distributions - Use Percentile. Any value beyond 2.5th percentile or 97th percentile is considered as Outlier.

- <b> How to treat outliers?
- After detecting outliers using Z Score, IQR Based Treatment, Percentile, Winsorization, we take any of the below methods to treat outliers.
- 1. <b>Trimming - Remove the records with outliers.
     Use when the number of outlers are less.
  2. <b>Capping (winsorization - only when we use Percentile technique ) - Replace those outliers with extreme left or right value. Ex: If we use Z score, and if Mean-3SD is 80, replace the outliers on the left extrement with 80.
     Use when there are more outliers.
  4. Treat it like a missing value, make it NaN
  5. Discretization - Age - 10-20, 20-30, etc. Make an outlier come to the range.
