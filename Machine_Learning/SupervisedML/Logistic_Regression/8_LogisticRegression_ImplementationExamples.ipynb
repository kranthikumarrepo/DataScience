#### Insurance.csv - only one input attribute(Single independent attribute)
- File has age(Independent variable) and Insurance data(Dependent variable), based on age predict whether a customer buys an insurance or not.

```python
# Read the required data

import pandas as pd

Insurance_Data = pd.read_csv(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\insurance_data.csv")
Insurance_Data.head(3)
```

```python
# Plot the scatterplot to see the data

import matplotlib.pyplot as plt

plt.xlabel("Age of Customer")
plt.ylabel("Bought Insurance")
plt.scatter(Insurance_Data.age,Insurance_Data.bought_insurance, marker = '+',color = 'r')
```

```python
# Calculate the decision boundary using the sigmoid function formula and draw the scatterplot and sigmoid function

import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

# Calculate sigmoid(x) = 1/(1 + e power -x)
def sigmoid(x):
  return 1/(1 + np.exp(-x))

x = Insurance_Data['age']
y = Insurance_Data.bought_insurance

# Send each value of x and calculate the value of y and store in a list.
mymodel = list(map(sigmoid, x))
#print(mymodel.sort)

# Now draw the scatter plot with the input values
plt.scatter(x, y)

# Draw the line plot using the input of x and derived y values.
plt.plot(x, mymodel)
plt.show()
```

```python
# Prepare the data into train and test data

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(Insurance_Data[['age']],Insurance_Data.bought_insurance,test_size = 0.1)
# for train_test_split, first parameter has to be multi dimensional array
```

```python
#Insurance_Data[['age']]
#Insurance_Data.bought_insurance
```

```python
# Train the model

from sklearn.linear_model import LogisticRegression

# Create object for LogisticRegression class
my_model = LogisticRegression()

# Train the model
my_model.fit(x_train,y_train)
```

```python
# Predict the outcome by giving test input data

y_predict = my_model.predict(x_test)
```

```python
# Probablity of yes or no for each test input data

my_model.predict_proba(x_test)
```

```python
# Check the score - Accuracy of the model

my_model.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_predict))
```

#### diabetes_data - multiple input attributes(Multiple independent attributes)
- File has multiple diabetes related parameters(Multiple independent variables), train the data with those parameters and predicut the outcome of diabetes risk.

```python
# Read the required data

import pandas as pd

diabetes_data = pd.read_csv(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\data_new.csv")
diabetes_data.head(5)
```

```python
# Get the independent columns and dependent column separately

# diabetes_data_x = diabetes_data.drop('diabetes_risk',axis=1)
diabetes_data_x = diabetes_data[['Age',	'Glucose',	'BMI',	'Insulin',	'BloodPressure',	'SkinThickness',
                                 'DiabetesPedigree',	'Pregnancies','Triglycerides',	'HDL']]

diabetes_data_y = diabetes_data['diabetes_risk']
#diabetes_data_y = diabetes_data[['diabetes_risk']]
```

```python
# Predict the outcome by giving test input data

y_predict = my_model.predict(x_test)
```
