### Classification algorithms
- Logistic regression is a supervised learning algorithm used for binary classification.
- Deals with Categorical Data
- Categorical Data - Classification algorithms in machine learning are supervised learning methods used to categorize data into predefined classes or categories.
- They learn from labeled data to predict the class of new, unseen data points.
- Some Common Classification Algorithms below:
- .
- Logistic Regression:
- Logistic regression is a supervised learning algorithm used for binary classification.
- A linear model that uses a sigmoid function to predict the probability of a data point belonging to a specific class.
- .
- Decision Tree:
- A tree-like structure that makes decisions based on feature values to classify data.
- .
- Random Forest:
- An ensemble method that combines multiple decision trees to improve accuracy and prevent overfitting.
- .
- K-Nearest Neighbors (KNN):
- Classifies data points based on the majority class of their k nearest neighbors in the feature space.
- .
- Neural Networks:
- Complex models inspired by the structure of the human brain, capable of learning intricate patterns for classification.

#### Logistic Regression (Logistic Classification)
- Prediction value is categorical
- Logistic regression is a supervised machine learning algorithm used for binary classification.
- Ex: email is spam or not, which category does a case fall in, will customer buy insurance or not.

#### Logistic Classification Types
- Binary Classification - Yes/No, True/False, etc.
- Multiclass Classification - Ex: Which party do you vote - Republic, Democratic, Independent, etc.

#### Note:-
- To use Logistic Regression - data should be linearly separable, data should be on extreme ends.
- Logistic Regression uses 2 types of algorithm in background to come with decision boundary (Like Regression line linear regression)
- 1. Perceptron trick
- 2. Sigmoid function - Mostly used.
- Note:-Regression line here is called decisoin boundary

#### Formula
- sigmoid(z) = 1/(1 + e power -z), where e = Eulers number = 2.71828 >>> Outcome of Sigmoid function is always between 0 and 1
- Decision boundary has value 0.5
- Values < 0.5 >>> Negative class
- Values > 0.5 >>>> Positive class
- Sigmoid(z) = 1/(1+e power z) = wi = w0-n(y-yx)x
