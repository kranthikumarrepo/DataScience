### Supervised Machine Learning Types

#### 1. Linear Regression - Numerical

Linear Regression Types - Predicted value is continues - Numerical

i. Simple Linear Regression - One input column and one output column

Steps involved in Simple Linear Regression:-

    - Take the input, draw scatterplot, now draw a linear lines that passes as close as possible to all the data points. Calculate the best line using one of the Error/Loss functions (Mean Absolute Error(MAE) and Mean Square Error(MSE) - Gradient Descent)
    
    MAE = Sigma |Prediction i - Observed i|/N
    
    MSE = Sigma Sq of (Prediction i - Observed i)/N

    - Once the best line is identified among all the lines, with least error, calculate the m and c values for that line.

    - Now predict the other values using the formula
            y = mx + c (In Machine learning terminology y = wc + b), 
            where y is output to be predicted, 
            m = slope or gradient - how steep the line is, 
            x = input, 
            c = y intercept, the point where the line crosses the y axis.

Note: All the input columns are represented by x(Training data), output is represented by y(Prediction data).

Formula:- y = mx + c or y = mx + b



ii. Multiple Linear Regression - More than One input column and one output column

Steps involved in Multiple Linear Regression:-

    - Take the input columns, draw scatterplot, now draw a linear lines that passes as close as possible to all the data points 
    for each individual column. 
    Calculate the best lines for each input column using one of the Error/Loss functions (Mean Absolute Error(MAE) and 
    Mean Square Error(MSE) - Gradient Descent)
    
    MAE = Sigma |Prediction i - Observed i|/N >>> More detailed explanation in 00_ML_Notes_CostFunction notes.
    
    MSE = Sigma Sq of (Prediction i - Observed i)/N >>> More detailed explanation in 00_ML_Notes_CostFunction notes.

    - Once the best lines are identified for all the input columns, with least error, calculate the mi and c values for those lines.

    - Now predict the other values using the formula
            y = (mixi + mjxj + mzxz + ..... ) + c (In Machine learning terminology y = wc + b), 
            where y is output to be predicted, 
            m = slope or gradient - how steep the line is, for each of the input columns. This is calculated for each input column
            x = input value for each input column, 
            c = y intercept, the point where the line crosses the y axis. This is common for all the input columns.

Note: All the input columns are represented by x(Training data), output is represented by y(Prediction data).

Formula:- y = (m1x1 + m2x2 + m3x3 + .... ) +c

Example

x1=163.3,x2=31.6,x3=52.9

m1 = 0.04472952, m2 = 0.18919505, m3 = 0.00276111

c = 2.97906734

((0.04472952 * 163.3) + (0.18919505 * 31.6) + (0.00276111 * 52.9) + 2.97906734) = 16.408024255
