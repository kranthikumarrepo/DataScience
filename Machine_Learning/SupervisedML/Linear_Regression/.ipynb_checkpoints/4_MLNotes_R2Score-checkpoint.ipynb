### R2 Score
- Useful to calculate the accuracy for Linear Regression model only.

https://www.youtube.com/watch?v=bMccdk8EdGo

- R2 score - Tells if the regression line is correctly predicting the actual values.
- In machine learning, the R-squared (RÂ²) score, also known as the coefficient of determination, is a statistical measure used to evaluate the goodness of fit of a regression model.
- It indicates the proportion of variance in the dependent variable that is predictable from the independent variable(s).
- Essentially, it tells you how well the model's predictions match the actual data points.
- It ranges from 0 to 1 >>> 0 is not at all good, 1 is the best.

#### Steps to calculate the R2 Score:-

- First draw a mean line and calculate the difference between the actual data points and mean value for all the elements(i.e. variance of mean).
- Draw a regression line (best fit line), calculate the difference between the actual data points and regression value for all the elements(i.e. variance of regression line).
- Formula = (Variance(Mean) - Variance(Regression line)) / Variance(Mean)
- R2 is negative - Wrong prediction
- R2 is close to 1 - Right prediction
- Less R2 score means not a good prediction
- r = CoRelation measures the linear relationship between two variables with respect to direction and strength.
- r2 measures of how close each data point fits to the regression line. It tells how well the regression line predicts the actual values

#### When do we go with adjusted R2 score
- When we used unnecessary columns in the ML training, we use Adjusted R2, which ignores the unnecessary columns whille calculating the R2 score.
- Adjusted R-squared is a modified version of R-squared that adjusts for the number of predictors in a model.
- It penalizes the addition of irrelevant predictors, making it a more reliable measure of model fit than R-squared, especially when comparing models with different numbers of independent variables.

#### Calculate the R2 score using python

```python
from sklearn.metrics import r2_score
import numpy as np

# Example data
y_true = np.array([3, -0.5, 2, 7])  # Actual values
y_pred = np.array([2.5, 0.0, 2, 8])  # Predicted values

# Calculate R2 score
r2 = r2_score(y_true, y_pred)

print(f"R-squared score: {r2}")
```
