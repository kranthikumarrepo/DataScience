```python
# Give the above x and y data to train, test, split function to create a separate data sets for x,y train and test

from sklearn.model_selection import train_test_split

x_trainTVRadioNews,x_testTVRadioNews,y_train_sales,y_test_sales = train_test_split(x_train_data,y_train_data,test_size=0.2,random_state=42)
```

```python
# Predict the sales for test input data
y_predict_sales = my_model.predict(x_testTVRadioNews)

# Create a dataframe from predict sales for easy indexing later to use joins for test and predicted data
df_y_predict_sales = pd.DataFrame(y_predict_sales)

# Rename column names from 0 to y_predict_sales
df_y_predict_sales.rename(columns={0: 'y_predict_sales'}, inplace=True)
df_y_predict_sales.head(5)
```

```python
# Create a dataframe from test sales for easy indexing later to use joins for test and predicted data
df_y_test_sales = pd.DataFrame(y_test_sales)

# Rename column names from Sales to y_test_sales
df_y_test_sales.rename(columns={'Sales': 'y_test_sales'}, inplace=True)

# Reset the index as the earlier index includes last 20 indexes from the train test split function.
df_y_test_sales.reset_index(drop=True,inplace=True)
df_y_test_sales.head(5)
```

```python
# Join the test and predicted sales to see them side by side

df_test_predict_sales = pd.merge(df_y_predict_sales,df_y_test_sales,how='inner',left_index=True, right_index=True)
df_test_predict_sales.head(5)
```

```python
# Test the prediction by using Mean Absolute Error method - |predicted value - actual value|N
# Manually

MAE_Manual = sum(abs(df_y_predict_sales.y_predict_sales - df_y_test_sales.y_test_sales))/40
MAE_Manual
```

```python
# By using the formula, calculate the MAE, MSQE, RMSQE

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import r2_score

MAE_Formula = mean_absolute_error(df_y_predict_sales.y_predict_sales, df_y_test_sales.y_test_sales)
print("MAE ", MAE_Formula)

mse_value = mean_squared_error(df_y_predict_sales.y_predict_sales, df_y_test_sales.y_test_sales)
print("mse_value ",mse_value)

RMSE = root_mean_squared_error(df_y_predict_sales.y_predict_sales, df_y_test_sales.y_test_sales)
print("RMSE ",RMSE)

R2Score = r2_score(df_y_test_sales.y_test_sales, df_y_predict_sales.y_predict_sales)
print("R2Score ",R2Score)
```

```python
# Check if manually calculated value and the formula value for MAE is same

MAE_Formula - MAE_Manual
```
