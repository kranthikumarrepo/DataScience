### Dataset 1 - Use k means clustering to create different clusters based on Age and Income

```python
# Age - Income grouping

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
Age_Income_DF = pd.read_csv(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\income.csv")
Age_Income_DF.head(3)
```

```python
# Plot the data on a scatterplot
import matplotlib.pyplot as plt

plt.scatter(Age_Income_DF.Age,Age_Income_DF['Income($)'])
```

```python
# Prepare the data to train the model

x = Age_Income_DF[['Age','Income($)']]
```

```python
# Create the model object
# Initially start with 3 clusters

from sklearn.cluster import KMeans
my_model = KMeans(n_clusters=3,random_state=42)
my_model
```

```python
# Train the model

my_model.fit(x)
```

```python
# Check the predicted value by using labels_ method
# This tells, which record falls to which cluster
# Since we have cluster size as 3, it creates 3 clusters 0,1,2

y_predicted = my_model.labels_
y_predicted
```

```python
# One more way is we can fit and predict at the same time
# Shows the same result as above

y_predicted = my_model.fit_predict(x)
y_predicted
```

```python
# We will append the predicted values to our Age_Income_DF dataframe for visualization

Age_Income_DF['y_Predicted'] = y_predicted
```

```python
# Plot the age, income and predicted value
# Different color for different cluster

Age_Income_DF_c1 = Age_Income_DF[Age_Income_DF.y_Predicted == 0]
Age_Income_DF_c2 = Age_Income_DF[Age_Income_DF.y_Predicted == 1]
Age_Income_DF_c3 = Age_Income_DF[Age_Income_DF.y_Predicted == 2]

plt.scatter(Age_Income_DF_c1.Age,Age_Income_DF_c1['Income($)'],color='red')
plt.scatter(Age_Income_DF_c2.Age,Age_Income_DF_c2['Income($)'],color='green')
plt.scatter(Age_Income_DF_c3.Age,Age_Income_DF_c3['Income($)'],color='black')

plt.xlabel('Age')
plt.ylabel('Income $')
plt.legend()
```

```python
# Plot using seaborn
# This can be done directly without needing to create 3 new dataframes based on y_Predicted as we did using matplotlib above.
import seaborn as sns
sns.scatterplot(x=Age_Income_DF.Age,y=Age_Income_DF['Income($)'],hue=Age_Income_DF.y_Predicted,data=Age_Income_DF)
```

```python
# Both the graphs are exactly same
# But the grouping is not proper as the data is not scaled, i.e. x axis age has values only from 27 to 42, whereas y axis income is 
# from 40,000 to 1,60,000
# Hence we need to scale the data, i.e. both Age and Income should be on same scale

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(x[['Age']])
x[['Age']] = scaler.transform(x[['Age']])

scaler.fit(x[['Income($)']])
x[['Income($)']] = scaler.transform(x[['Income($)']])
x.head(3)
```

```python
# Now train the model with scaled data

y_predicted = my_model.fit_predict(x)
x['y_Predicted'] = y_predicted
```

```python
# To find the centroids

centroids = my_model.cluster_centers_
```

```python
# Plot using seaborn

import seaborn as sns
plt.scatter(centroids[:,0],centroids[:,1],color='purple',marker='+',label='Centroid')
sns.scatterplot(x=x.Age,y=x['Income($)'],hue=x.y_Predicted,data=Age_Income_DF)
```

```python
# Now using SSE and Elbow method, find the ideal k value

# Initialize the Sum of Squared Error
sse = []
for i in range(1,10):
    my_model = KMeans(n_clusters=i,random_state=42)
    my_model.fit(x)
    sse.append(my_model.inertia_)
sse
```

```python
# Plot the Elbow graph using sse score and k value

plt.xlabel('K Value')
plt.ylabel('SSE Score')
plt.plot(range(1,10),sse)
```

```python
# Looking at the plot, 3 seems like the correct value for k and hence we need not train the model again, as we already trained the model
# using k value as 3
```

### Dataset 2 - Use k means clustering to create different clusters for different data using the below columns

```python
# Import the file

import pandas as pd

ConstructionPermits_DF = pd.read_csv(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\main_data.csv")
ConstructionPermits_DF.head(5)
```

```python
# Get the columns required for k means clustering 

#x = ConstructionPermits_DF.iloc[:,1:]
x = ConstructionPermits_DF[['fy2017, Dealing with construction permits','fy2019, Dealing with construction permits',	
                            'fy2018, Dealing with construction permits']]
x.head(3)
```

```python
# Create k means cluster algorithm, initially start with 3 clusters

from sklearn.cluster import KMeans

# Initialize the number of clusters
my_model = KMeans(n_clusters=3,random_state=42)
```

```python
# Train the model
# my_model.fit(x)

# Predict the outcome
# my_model.labels_

# Train the model and predict the outcome in one ste.
my_model.fit_predict(x)
```

```python
# Calculate WCSS Score for 3 clusters

my_model.inertia_
```

```python
# Create a for loop to iterate the above process so that we get different WCSS values for each k value, used to draw Elbow graph
# Initilize empty list for WCSS Values

WCSS = []

# Iterate the cluster size from 1 to 15
for i in range(1,16):
    my_model = KMeans(n_clusters=i,random_state=42)
    my_model.fit(x)

    # Store the value of WCSS for each cluster size
    WCSS.append(my_model.inertia_)
WCSS
```

```python
# Plot the graph using the k values on x axis and WCSS Value on Y Axis
import matplotlib.pyplot as plt

plt.xlabel("Cluster Size - k")
plt.ylabel("WCSS Value")
plt.plot(range(1,16),WCSS)
```

```python
# Based on the above elbow graph, lets take cluster size as 5 and build the model.

my_model = KMeans(n_clusters=4,random_state=42)
my_model.fit(x)
```

```python
# WCSS Score

my_model.inertia_
```

```python
# Check the predicted value by using labels_ method
# This tells, which record falls to which cluster
# Since we have cluster size as 7, it creates 7 clusters 0,1,2,3,4,5,6

y_predicted = my_model.labels_
y_predicted

# 1st row is assigned to 4th cluster
# 2nd row is assigned to 3rd cluster
```

```python
# Centroids

my_model.cluster_centers_

# Since this is based on 3 input attributes, the below are the Centroid positions of 3D array
```

```python
# As we increase the cluster size, the WCSS Score might decrease, but that is not properly categorized.
# We should decide the cluster size based on elbow diagram and customer requirement.
```

```python
# Measure the accuracy of the prediction.
# silhouette_score for K Means Clustering.

from sklearn.metrics import silhouette_score
sscore = silhouette_score(x,y_predicted)
sscore
# silhouette_score with clustersize = 3 is 0.555, but still number of clusters should be decided on the elbow chart not on this score
```
