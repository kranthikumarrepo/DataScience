#### Concept of SVM:-


- All the datapoints are classified into +ve OR -ve classes in classification problems. The separator is called as decision boundary.
- If the datapoint is close to the decision boundary line(hyperplane), it means the probablity of that datapoint to be in the +ve or -ve hyperplane is less.
- If the datapoint is far away from the decision boundary line(hyperplane), it means the probablity of that datapoint to be in the +ve or -ve hyperplane is high. 
- In the above diagram, SVM tries to identify the hyperplane that tries to classify the datapoints, as widely as possible. Ex: Here π1 is selected.
- <b>SVM :- </b>Multiple decision boundary lines are drawn and the decision boundary line(hyperplane) that is maximum far away from both the +ve and -ve datapoints is selected, so that in future a new datapont comes, it should be able to accomodate that datapoint within the positive or negative classes.

#### How a decision boundary line is selected among multiple decision boundary lines
- 1. First randomly select a hyperplane(decision boundary line) - π(pi) - This will classify the datapoints into +ve and -ve classes.
- 2. Go in the +ve (Green) direction, parallel to π, until we reach the first datapoint in green - This point is called as<b> +ve Support Vector</b>. Stop here and draw a hyperplane(decision boundary line), this can be called as π +ve.
- 3. Now go in the -ve(Red) direction, parallel to π, until we reach the first datapoint in red - This is called as<b> -ve Support vector</b>. Stop here and draw a hyperplane(decision boundary line), this can be calleda as π -ve.
- 4. Next calculate the shortest distance between π +ve and π - ve, we call this as Margin ( d ).
- 5. Now, repeat the steps (1 to 4) for different hyperplanes ( π )
- 6. Select the hyperplane ( π ) where the d value is higher compared to all the other hyperplanes.

## Decision Rule
- Once the decision boundary is identified, how do we classfy a new data whether it is on the +ve side or -ve side of the decision boundary?
- Decision Rule will help us to classify new data.
- .
- When an unknow data point is given, we need to identify if that point is on the -ve or +ve side of the decision boundary.
- Step1:- Create a vector(w) perpendicular to decision boundary(Or +ve Support Vector/ -ve Support Vector). For this vector, we only know about the direction, but we do not know the length of this vector. This vector at one point meets the decision boundary.
- Step2:- Take co-ordinates of the new data point (x,y), which is also a vector(u).
- Step3:- Project vector u on vector w. Project means keep one over the other. Since projection is a dot product, so the equation is w.u
- Assumption:- Let us consider the distance from origin to the decision boundary as c.
- Step4:- Once we projecct vector u on vector w, the dot product will be w.u & this will be either equal to/ greater than the distance between origin and decision, so the equation is w.u >= c
- We can change the equation from
- w.c >= c >>>  w.c - c >= 0 >>> w.c + b >= 0
- Step5:- Once we apply the equation, if the value is +ve, then the new datapoint is on +ve side and if -ve, it is on -ve point.

### Find the distance (d) between +ve and -ve decision boundary
### Constraints for Green and Red points i.e. the existing points.
- The distance d we derive should be as much maximum as possible while making sure that the green and red (existing data points) should always be above or below the +ve and -ve decision boundaries. The distance should be maximum because that makes the SVM robust to the new points.

- The green and red points, which are above and below +ve and -ve decision boundaries, should always be above or below those decision boundaries.
- So the equation for Green points is
- w.x + b >= 1
- and the equation for Red points is
- w.x + b <= -1
- .
- Convert above two equations into 1 equation for simplification.
- Label(y) all +ve points with +1 and all the -ve points with label -1
- So the equation w.x + b >= 1, when multiplied with +1 label y, changese to y(w.x + b) >= 1
- and when we multiply w.x + b <= -1, with -1 label y, changes to y(w.x + b) >= 1
- So, the equation changes to y(w.x + b) >= 1 for all the vectors
- and only for the support vectors the equation is y(w.x + b) = 1
- .
- Now calculate the distance d, while the equation y(w.x + b) >= 1, which means all the +ve and -ve points are above or below the /+ve or -ve support vectors.
- Enter a loop
- Keep increasing the distance, until the equation is valid y(w.x + b) >= 1, which means, until the decision boundaries have not crossed the +ve and -ve points.
- https://youtu.be/yCAlHPDgWtM?si=58YVOtWyYi2i7Qtd&t=1632

- .
- So, to calculate the distance d, take two support vectors, one from +Ve and the other from -Ve decision boundary, and then calculate the distance between those two vectors, which is (x2-x1)
- Then draw a line perpendicular to decision boundary, which will be vector w. To make vector w to scaler, take magnitude, i.e. w/||w||
- Now project (x2-x1) vector on unit vector w, which will be dot product
- So, the equation of dot product between (x2-x2) vector and unit vector w is (x2-x1).w/||w||
- So, this perpendicular vector will be same as distance d
- So, d = (x2-x1).w/||w|| = (x2.w - x1.w) / ||w||
- .
- For support vectors:-
- y(w.x + b) = 1 >>>
- For green support vector(x = x2), y will be = 1
- y(w.x + b) >= 1 >>> 1(w.x2 + b) = 1 >>> w.x2 = 1-b
- For red support vector(x = x1), y will be = -1
- y(w.x + b) >= 1 >>> -1(w.x1 + b) = 1 >>> w.x1 = -b-1
- 
- So, d = (x2.w - x1.w) / ||w|| changes to >>> ((1-b) - (-b-1)) / ||w|| >>> (1-b+b+1)/||w|| >>> 2/||w||
- So<b> d = 2/||w||
- So the final equation - optimization function to derive distance d while keeping the greeen and red points above/below -ve and +ve decision boundaries is
- <b>argmax (w.b) = 2 / ||w||, while y(w.x + b) >= 1

## Hard Margin SVM
- Above equation is called Hard Margin SVM i.e. to find the max distance d, we are assuming the green and red points are above and below their respective +ve and -ve decision boundaries, even if 1 point enters into the other decision boundary, we will not be able to find the distance d.
- i.e. the data should be linearly seperable.
- 
- From above pic, we can see, even if 1 green point falls in -ve plane, the distance d cannot be calculated, hence this is called Hard Margin SVM, which only works in the case of Perfectly Linearly Seperable data.

### ReWriting the equation of maximizing function / Error function - Margin Error.
- The below equation is where we are trying to maximize the margin of the decision boundaries.
- Max f(x) = argmax (w.b) = 2 / ||w||
- Max f(x) = 1/ (Min f(x)) >>> Min f(x) = 1/Max f(x)
- So, Min f(x) = argmin (w.b) = ||w|| / 2

### Zeta value for all the points on the plane - Classificatoin Error.
- Helps to minimize the error for incorectly classified items.
- Σi=1 to n Ζi >>> i = each data point
- For all the correctly classified points,  Ζi is zero.

- From above diagram, all the green points above +ve decision boundary, the  Ζi is 0 and for all the red points below the -ve decision boundary, the  Ζi is 0.
- But, for incorrectly classified items like circled green data points below decision boundary and circled red data points above decision boundary, the  Ζi is non zero >>> Calculation of Zi is shown below.
- <b> Calculation of Zi
- It is the distance between the data point and their respective +ve or -ve decision boundary. i.e. for green circled data points, Zi is the distance between the data point and the +ve (pi +ve) decision boundary and for red circled data points, Zi is the distance between the data point and the -ve (pi -ve) decision boundary.

## Support vector machine has two error/cost functions
- SVM Error function is a combination of 2 parameters
- .
1. <b>Margin error - Helps to maximize the distance between decision boundaries.
    - Margin error is inversly proportional to margin. When the margin is less, the errors are more. When the margin is more, the errors are less.
    - Equation:- Max f(x) = argmax (w.b) = 2 / ||w||
    - .
3. <b>Classification error >>> Controlled using C hyper parameter
    - Helps to minimize the error for incorectly classified items.
    - Properly classify the +ve and -ve hyperplanes using the +ve and -ve decision boundaries. i.e. no data points should exist
in between +ve and -ve decision boundaries.
    - Equation:- Σi=1 to n Ζi >>> i = each data point
    - Smaller the number, wider the +ve and -ve decision boundaries and some data points exist after +ve and -ve decision boundaries.
    - Higher the number, smaller the distance between +ve and -ve decision boundaries, and higher the correctness of classification - Perfectly classify the +ve and -ve items.


## Soft Margin SVM
- Here we try to accomodate some misclassification (errors), by adjusting hyper parameter C, hence the name Soft Margin SVM.
- Earlier we used only margin error while classifying the data points, now we will also use classification error.
- .
- If we use only margin error, it will classify all the items correctly, not even with one error. But if we use classification error, it helps us to widen the margin, but it can accomodate more erros.
- Margin Error will find the max distance d, will not work in case of non-linear data. i.e. optimization function to derive distance d while keeping the green and red points above/below -ve and +ve decision boundaries.
- .
- <b> argmax (w.b) = 2 / ||w||, while every data point should always be >= 1 i.e. y(w.x + b) >= 1 >>>>>>>>>>> This will not work for non-linear data.
- Ex: when we take a red point above +ve hyperplane, the equation would be label = -1 as it is a red point, and the equation w.x + b = +ve
- so -1 (+ve) >= 0 is false.
- This problem is called constrain optimization problem.
- .
- <b>In Soft Margin SVM, we will be able to handle outliers, non-linear data.
- Most real world data is not perfectly linear.
- .
- <b>Derivation of Softmargin SVM :- Margin Error + C * Classification Error
- Min f(x) = argmin (w.b) = ||w|| / 2 + C . Σi=1 to n Ζi

- Here in the formula, the first part is trying the expand the decision boundary (Margin error) and the second part is trying to minimize the errors(Classification error).
- .
- <b> Significance of hyper parameter C
- When we increase the value of C, we are trying to give more weightabe to Classification, which means, margin is secondary importance, but give more importance to classification, i.e. focus more on correctly classiying the data points.
- .
- 
- The above diagram (C Value is more - focus more on classiificatoin), shows that all the items are correctly classified, but the margin is less.
- 
- The above diagram (C value is less - focus more on margin), shows that all the items are not correctly classified, but the margin is more.
- So, the ideal value of C depends on our requirement. This can also be decided after performing Grid Search CV, cross validation CV.
