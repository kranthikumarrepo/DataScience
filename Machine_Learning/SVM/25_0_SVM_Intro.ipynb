#### Concept of SVM:-


- All the datapoints are classified into +ve OR -ve classes in classification problems. The separator is called as decision boundary.
- If the datapoint is close to the decision boundary line(hyperplane), it means the probablity of that datapoint to be in the +ve or -ve hyperplane is less.
- If the datapoint is far away from the decision boundary line(hyperplane), it means the probablity of that datapoint to be in the +ve or -ve hyperplane is high. 
- In the above diagram, SVM tries to identify the hyperplane that tries to classify the datapoints, as widely as possible. Ex: Here 1 is selected.
- <b>SVM :- </b>Multiple decision boundary lines are drawn and the decision boundary line(hyperplane) that is maximum far away from both the +ve and -ve datapoints is selected, so that in future a new datapont comes, it should be able to accomodate that datapoint within the positive or negative classes.

#### How a decision boundary line is selected among multiple decision boundary lines
- 1. First randomly select a hyperplane(decision boundary line) - (pi) - This will classify the datapoints into +ve and -ve classes.
- 2. Go in the +ve (Green) direction, parallel to , until we reach the first datapoint in green - This point is called as<b> +ve Support Vector</b>. Stop here and draw a hyperplane(decision boundary line), this can be called as  +ve.
- 3. Now go in the -ve(Red) direction, parallel to , until we reach the first datapoint in red - This is called as<b> -ve Support vector</b>. Stop here and draw a hyperplane(decision boundary line), this can be calleda as  -ve.
- 4. Next calculate the shortest distance between  +ve and  - ve, we call this as Margin ( d ).
- 5. Now, repeat the steps (1 to 4) for different hyperplanes (  )
- 6. Select the hyperplane (  ) where the d value is higher compared to all the other hyperplanes.

## Equation of a Hyper-plane in N dimensions
- To derive the equation of hyperplane, start with a 2D Line equation.
- 2D Line Equation
- 
- 2D Line equation >>> y = mx + b
- Above equation is popular, but the general form to the above equation is ax + by + c = 0
- So we can say m = -a/b and b = -c/b >>> y= -a/b*x -c/b
- <b>ax + by + c = 0 >>> This is the general form of a linear equation, representing a straight line.
- .
- To make the above equation more simple, let us call x axis as x1 and y axis as y1,
- 
- so the equation changes to ax1+bx2+c=0 >>> 2D
- If we are working in 3D, then the equation is ax1 + bx2 + cx3 + c = 0
- If we are working in 4D, then the equation is ax1 + bx2 + cx3 + dx4 + c = 0
- But if we are working with more than 26 features, then above equation cannot be represented, hence we replace a, b, c and so on with w1, w2, w3, and so on and constant c with w0.
- So the above equation for 2D Line:- ax + by + c = 0 changes to
- .
- <b> Equation of Line >>> w1x1 + w2x2 + w0 = 0
- .
- 
- If we are working in 3D(Above diagram), then the 
- <b> Equation of Plane >>> w1x1 + w2x2 + w3x3 + w0 = 0

- <b> Equation of Hyperplane in 4D >>> w1x1 + w2x2 + w3x3 + w4x4 + w0 = 0
- .
- <b>For N dimension, the
- <b> Equation of HyperPlane in N Dimension >>> w1x1 + w2x2 + w3x3 + ..... wNxN + w0 = 0, where w0 = intercept
- .

### To further simplify the above equation - Equation for N dimension hyperplane.

- 
- Equation of line :- w1x11 + w2x2 + w0 = 0
- 
- If there was no intercept (Origin), then the equation would change to w1x1 + w2x2 = 0 >>> Also called as dot product.
- DOT PRODUCT:-
- To get the dot product of vector (x1,x2), take w1 and w2 as an vector (w1,w2). Above equation can be written as dot product [w1,w2] = w1x2 + w2x2
- For 2D :- Dot product [ w.x ] = w1x1 + w2x2
- For 3D :- Dot product [ w.x ] = w1x1 + w2x2 + w3x3
- For ND :- Dot product [ w,x ] = w1x1 + w2x2 + .... + wNxN
- So assume there are two vectors, Vector w with components [w1,w2,... wN] and Vector x with components [x1,x2,...xN]
- So we can reduce the above final equation for N dimension w1x1 + w2x2 + w3x3 + ..... wNxN + w0 = 0
- to 
- w.x + w0 = 0 (W dot product of x + w0 = 0)
- Transpose :- Interchanging rows and columns of a matrix.
- w.x (w dot product of x) >>> wTx (w transpose x)
- 
- So the final equation of w.x + w0 = 0
- is
- wTx + w0 = 0

- In the above equation, make w0 (Intercept/ Origin) = 0, which means we are making the line pass through Origin. See below pic.

- So the final equation wTx + w0 = 0
- changes to
- wTx = 0 (Assuming intercept/Hyperplane is passing through origin)
- So for
- 2D :- wTx >>> w1x1 + w2x2 = 0
- 3D :- wTx >>> w1x1 + w2x2 + w3x3 = 0
- For ND :- wTx >>> w1x1 + w2x2 + w3x3 + ... + wNxN = 0
