```python
# Import the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.axes._axes import _log as matplotlib_axes_logger
from mpl_toolkits import mplot3d
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
from sklearn.metrics import accuracy_score
```

```python
# Generate non-linear data & plot the data.

from sklearn.datasets import make_circles
X, y = make_circles(100, factor=.1, noise=.1)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='bwr')
```

```python
# Prepare the data for model.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

### Use Kernel = linear for Kernel transformation - doesn't work for non-linear data.

```python
# Create SVC object using Linear Kernel - This will not give proper results as the data is non-linear but we Kernel = linear which only works
# for linear data.

classifier = SVC(kernel="linear")
# Default is rbf

#classifier.fit(X_train, y_train.ravel())
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)

# We can see the accuracy score is just 0.55
```

```python
# Create a generalized function to plot the data and decision boundary

zero_one_colourmap = ListedColormap(('blue', 'red'))
def plot_decision_boundary(X, y, clf):
    X_set, y_set = X, y
    X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, 
                                 stop = X_set[:, 0].max() + 1, 
                                 step = 0.01),
                       np.arange(start = X_set[:, 1].min() - 1, 
                                 stop = X_set[:, 1].max() + 1, 
                                 step = 0.01))
  
    plt.contourf(X1, X2, clf.predict(np.array([X1.ravel(), 
                                             X2.ravel()]).T).reshape(X1.shape),
               alpha = 0.75, 
               cmap = zero_one_colourmap)
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = (zero_one_colourmap)(i), label = j)
    plt.title('SVM Decision Boundary')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
    return plt.show()
```

```python
# Plot the decision boundary generated using Kernel = Linear

plot_decision_boundary(X, y, classifier)
# We can see the decision boundary is not properly classfying the data.
```

```python
# Use Kernel = RBF >>> Here instead of using the sklearn, we are transforming the data from lower to higher dimension manually using the formula.
# Formula of Kernel Function(RBF):- f(x) = e power -x square >>> np.exp(-(X ** 2).sum(1))
# 2D has become 3D and this enables us to easily classify.

def plot_3d_plot(X, y):
    r = np.exp(-(X ** 2).sum(1))
    # This is a radial basis function - rbf applied on x and y co-ordinate, then sum new x and y co-ordinates and then keep that sum in r, which is 
    # plotted on z co-ordinate.
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:, 0], X[:, 1], r, c=y, s=100, cmap='bwr')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('y')
    return ax
```

```python
plot_3d_plot(X,y)
```

```python
# From above function, if we remove Z axis, i.e. r value i.e. rbf value, then everything is in 2D and data is still at center.

def plot_3d_plot(X, y):
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:, 0], X[:, 1], c=y, s=100, cmap='bwr')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('y')
    return ax
```

```python
plot_3d_plot(X,y)
```

### Use sklearn, inbuilt - Kernel = rbf for Kernel transformation - works for non-linear data.

```python
rbf_classifier = SVC(kernel="rbf")
# in background it uses the above formula for kernel transformation.

rbf_classifier.fit(X_train, y_train)
y_pred = rbf_classifier.predict(X_test)
accuracy_score(y_test, y_pred)

# Now we can see the accuracy score has been improved to 1
```

```python
plot_decision_boundary(X, y, rbf_classifier)

# We can see the decision boundary is properly seperating the data points.
```

### Use Kernel = poly for Kernel transformation - works for non-linear data.

```python
poly_classifier = SVC(kernel="poly",degree=2)
# degree parameter is used only with poly kernel.

poly_classifier.fit(X_train, y_train)
y_pred = poly_classifier.predict(X_test)
accuracy_score(y_test, y_pred)

# When degree = 3, accuracy score = 0.6
# When degree = 2, we can see the accuracy score has been improved to 1
```

```python
plot_decision_boundary(X, y, poly_classifier)

# We can see the decision boundary is properly seperating the data points.
```
