```python
# Get the required file data

import pandas as pd

SNA_DF = pd.read_csv(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\Social_Network_Ads.csv").drop(columns='User ID',axis=1)
SNA_DF.head(2)
```

```python
# Prepare the data

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(SNA_DF.drop(columns='Purchased',axis=1),SNA_DF.Purchased)
y_train.head(2)
```

```python
x_test['Gender'].head(2)
```

```python
# Encode the categorical data.

from sklearn.preprocessing import OneHotEncoder

OneHotEncoder_Obj = OneHotEncoder(drop='first',sparse_output=False)
x_train_Gender_Encoded = OneHotEncoder_Obj.fit_transform(x_train[['Gender']])
x_test_Gender_Encoded = OneHotEncoder_Obj.transform(x_test[['Gender']])
```

```python
OneHotEncoder_Obj.get_feature_names_out()
```

```python
x_train_Gender_Encoded[0]
```

```python
x_train_Gender_Encoded_DF = pd.DataFrame(x_train_Gender_Encoded,columns=['Gender_Male'])
x_test_Gender_Encoded_DF = pd.DataFrame(x_test_Gender_Encoded,columns=['Gender_Male'])
```

```python
# Scale the numerical data

from sklearn.preprocessing import StandardScaler

StandardScaler_Obj_Age_EstiSal = StandardScaler()
StandardScaler_Obj_Age_EstiSal.fit(x_train[['Age','EstimatedSalary']])
```

```python
x_train_Age_EstiSal_Scaled = StandardScaler_Obj_Age_EstiSal.transform(x_train[['Age','EstimatedSalary']])
x_test_Age_EstiSal_Scaled = StandardScaler_Obj_Age_EstiSal.transform(x_test[['Age','EstimatedSalary']])
```

```python
x_train_Age_EstiSal_Scaled[0]
```

```python
StandardScaler_Obj_Age_EstiSal.get_feature_names_out()
```

```python
x_train_Age_EstiSal_Scaled_DF = pd.DataFrame(x_train_Age_EstiSal_Scaled,columns=['Age_Scaled','Estimated_Scaled'])
x_test_Age_EstiSal_Scaled_DF = pd.DataFrame(x_test_Age_EstiSal_Scaled,columns=['Age_Scaled','Estimated_Scaled'])
```

```python
x_train_Age_EstiSal_Scaled_DF.head(2)
```

```python
# Prepare the scaled and encoded data

x_train_transformed = x_train_Age_EstiSal_Scaled_DF
x_train_transformed['Gender_Encoded'] = x_train_Gender_Encoded_DF
```

```python
x_test_transformed = x_test_Age_EstiSal_Scaled_DF
x_test_transformed['Gender_Encoded'] = x_test_Gender_Encoded_DF
```

```python
# Use the SVC model to classify the data

from sklearn.svm import SVC

# Create an SVM classifier with an RBF kernel
model = SVC(kernel='rbf', C=1.0, gamma='scale', degree = 9)
```

```python
# Train the model
model.fit(x_train_transformed, y_train)
```

```python
# Make predictions
y_pred = model.predict(x_test_transformed)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

### Using column Transformer
- To perform encoding and scaling in a single step

```python
x_train.head(2)
```

```python
from sklearn.compose import ColumnTransformer

ColumnTransformer_Obj = ColumnTransformer([
    ('Trans_Gender',OneHotEncoder(drop='first',sparse_output=False),['Gender']),
  ('Trans_Age_EstSal',StandardScaler(),['Age','EstimatedSalary'])
],remainder='passthrough')
```

```python
ColumnTransformer_Obj.fit(x_train)
```

```python
x_train_transformed = ColumnTransformer_Obj.transform(x_train)
```

```python
x_test_transformed = ColumnTransformer_Obj.transform(x_test)
```

```python
from sklearn.svm import SVC

# Create an SVM classifier with an RBF kernel
model = SVC(kernel='rbf', C=1.0, gamma='scale')
```

```python
# Train the model
model.fit(x_train_transformed, y_train)
```

```python
# Make predictions
y_pred = model.predict(x_test_transformed)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
