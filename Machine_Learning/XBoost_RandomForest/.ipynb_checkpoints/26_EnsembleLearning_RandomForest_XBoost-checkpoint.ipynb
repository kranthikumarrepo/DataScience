```python
import pandas as pd

#DF = pd.read_csv(r'C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\.csv')
#DF.head(1)
```

```python
from sklearn.datasets import load_iris

x,y = load_iris(return_X_y = True)
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

RandomForestClassifier_Obj = RandomForestClassifier(n_estimators=2)
```

```python
RandomForestClassifier_Obj.fit(x_train,y_train)
```

```python
y_predict = RandomForestClassifier_Obj.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_predict,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

DecisionTreeClassifier_Obj = DecisionTreeClassifier()
DecisionTreeClassifier_Obj.fit(x_train,y_train)
```

```python
y_predictDT = DecisionTreeClassifier_Obj.predict(x_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_predict,y_test)
```
