```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Load IRIS Dataset

X,y = load_iris(return_X_y=True)
# Data = load_iris(return_X_y=False)
# X = Data.data
# y = Data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Scaling

StandardScaler_Obj = StandardScaler()
StandardScaler_Obj.fit(X_train)
X_train_Transformed = StandardScaler_Obj.transform(X_train)
X_test_Transformed = StandardScaler_Obj.transform(X_test)
print("Train shape ", X_train_Transformed.shape, " Test Shape ",X_test_Transformed.shape)
```

```python
from sklearn.neural_network import MLPClassifier

MLPClassifier_Obj_S = MLPClassifier()
MLPClassifier_Obj_S.fit(X_train_Transformed,y_train)

MLPClassifier_Obj_NS = MLPClassifier()
MLPClassifier_Obj_NS.fit(X_train,y_train)
```

```python
y_Pred_S = MLPClassifier_Obj.predict(X_test_Transformed)
y_Pred_NS = MLPClassifier_Obj.predict(X_test)
```

```python
accuracy_score(y_Pred_S,y_test)
```

```python
accuracy_score(y_Pred_NS,y_test)
```

```python
print(classification_report(y_Pred_S,y_test))
```

```python
y_Pred_LR_NS = LogisticRegression_Obj_NS.predict(X_test)

y_Pred_LR_S = LogisticRegression_Obj_S.predict(X_test_Transformed)
```
