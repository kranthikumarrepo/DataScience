#### All functions
- To open a file
- Syntax:-
- file_pandas_dfname = pandas.read_csv / pandas.read_excel / etc.
- .
- To get info. of the file
- Syntax:-
- file_pandas_dfname.info() >>> Column names, data types, count of each column values, etc.
- .
- To get summary of the file
- Syntax:-
- file_pandas_dfname.describe() >>> Each column count, mean, SD, min, max, Q1,Q2,Q3 percentile.
- .
- To see all the columns in the file
- Syntax:-
- file_pandas_dfname.columns
- .
- To see values in a particular column
- file_pandas_dfname.column_name
- .
- To get the count of values in a particular column
- file_pandas_dfname.column_name.count()
- .
- To create our own index
- Syntax:-
- file_pandas_dfname.set_index('Existing_column_name', inplace = True)
- inplace >>> Makes the change permanant.
- Note: Once the column is made as index, it disappears from the columns list.
- .
- To remove the created index and make it as column again
- file_pandas_dfname.reset_index(inplace=True)
- .
- To Write To a file
- file_pandas_dfname.to_csv(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\MachineLearningFiles\InputCGPA_OutputPredictPackage.csv",index=False)
