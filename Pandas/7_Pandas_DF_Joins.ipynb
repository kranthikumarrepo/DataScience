### Joins
- inner, outer, left, right, self, cross
- .
- To concatenate two dataframes one below the other
- Syntax:-
- pandas.concat([df1_name,df2_name],axis=0)
- .
- To concatenate two dataframes side by side, like join
- Syntax:-
- pandas.concat([df1_name,df2_name],axis=1)
- .
- <b>Join syntax</b>:-
- pandas.merge(dataframe1Name,dataframe2Name,how='join type',on='column_name')
- Where
- merge = JOIN in Pandas
- how = type of join like inner, left, right, full, etc.
- on = column on which join condition is applied.

### Join
- join = merge in Pandas
- how = type of join like inner, left, right, full, etc.
- on = column on which join condition is applied.

```python
pd.merge(df1,df2,how='inner',on='ID')
```

```python
'''
how = inner, outer, left, right, self, cross
on = common column to join
merge = join
'''
```

```python
pd.merge(df1,df2,how='left',on='ID')
```

```python
pd.merge(df1,df2,how='right',on='ID')
```

```python
pd.merge(df1,df2,how='outer',on='ID')
```

```python
pd.merge(df1,df2,how='cross')
```
