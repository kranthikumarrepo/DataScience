### Graphs using kind in Pandas

- kind method in Pandas is a versatile tool for creating various types of plots directly from a DataFrame, leveraging Matplotlib as its backend.
- Easy to create the graphs.
- The kind argument specifies the type of plot to generate.
- Syntax
- dataframe.plot(x= Col1, y = Col2, kind='...') >>> On multiple columns
- dataframe.Column_Name.plot(kind='...') >>> On single column
- .
- Common values for the kind parameter include:
- 'line': Creates a line plot, often used to show trends over time.
- 'bar': Generates a vertical bar plot, suitable for comparing categorical data.
- 'barh': Produces a horizontal bar plot.
- 'hist': Displays a histogram, used to visualize the distribution of numerical data.
- 'box': Creates a box plot, useful for showing the distribution and outliers of numerical data.
- 'kde': Generates a Kernel Density Estimate plot, showing the probability density function of a continuous variable.
- 'density': An alias for 'kde'.
- 'area': Produces an area plot, similar to a line plot but with the area beneath the line filled.
- 'pie': Creates a pie chart, used to represent proportions of a whole.
- 'scatter': Generates a scatter plot, used to visualize the relationship between two numerical variables.
- 'hexbin': Creates a hexbin plot, useful for visualizing the density of points in a scatter plot when there are many data points.

#### Using dataframe.plot(x='col1',y='col2',kind='graph_type') >>> Multiple columns

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Category': ['A', 'B', 'C', 'D'],
        'Metric1': [20, 35, 30, 45],
        'Metric2': [15, 28, 25, 40]}
df = pd.DataFrame(data)

# Plotting 'Category' on x-axis and 'Metric1' on y-axis as a bar plot
df.plot(x='Category', y='Metric1', kind='bar')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Time': [1, 2, 3, 4, 5],
        'Value1': [10, 12, 15, 13, 18],
        'Value2': [8, 11, 14, 16, 17]}
df = pd.DataFrame(data)

# Plotting 'Time' on x-axis and 'Value1' on y-axis as a line plot
df.plot(x='Time', y='Value1', kind='line')
plt.title('Line Plot of Value1 over Time')
plt.show()
```

### Crosstab
- Syntax:-
- pandas.crosstab(df_obj_name.Col1,df_obj_name.Col2)
- .
- crosstab is a function used to compute a simple cross-tabulation of two or more factors (categorical variables).
- By default, it generates a frequency table, showing how often different combinations of these factors appear in the dataset.
- .
- Uses of crosstab:
- .
- Analyzing Relationships between Categorical Variables:
- crosstab is a fundamental tool for understanding the relationship between two or more categorical variables. For instance, in survey data, it can reveal the frequency of specific responses based on demographic categories like gender or age group.
- .
- Summarizing Data:
- It provides a concise summary of the joint distribution of variables, making it easier to see patterns and trends that might not be immediately obvious from raw data.
- .
- Data Validation and Consistency Checks:
- By examining the frequencies of combinations, crosstab can help identify inconsistencies or unexpected patterns in the data, aiding in data cleaning and validation.
- .
- Market Research and Social Sciences:
- It is widely used in these fields to explore relationships between variables, such as product preferences across different customer segments or opinions on social issues among various demographic groups.
- .
- Generating Contingency Tables:
- crosstab directly creates contingency tables, which are essential for statistical tests like chi-squared tests to determine the independence of categorical variables.
