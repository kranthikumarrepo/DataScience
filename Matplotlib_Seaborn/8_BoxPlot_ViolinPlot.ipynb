### Box Plot vs Violin Plot
- Box plots and violin plots are both ways to visualize the distribution of numerical data, but they offer different levels of detail.
- Box plots show the five-number summary (minimum, first quartile, median, third quartile, and maximum), while violin plots add a kernel density plot on each side of the box, revealing the distribution of the data and its density
- Box Plot tells if there are outliers.
- Violin Plot tells if the data is properly distributed or not.

```python
# Import tips dataset 

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings

tips_data = sns.load_dataset('tips')
```

#### Create a boxplot using Matplotlib
- This is used to identify the outliers.

```python
plt.boxplot(tips_data.tip)
warnings.filterwarnings('ignore')
```

#### Create a boxplot using Seaborn

```python
sns.boxplot(x = tips_data.tip, data = tips_data)
warnings.filterwarnings('ignore')
```

#### Create a Violin Plot Using Matplotlib

```python
plt.violinplot(tips_data.tip)
warnings.filterwarnings('ignore')
```

#### Create a Violin Plot Using Seaborn

```python
sns.violinplot(y = tips_data.tip, data = tips_data)
warnings.filterwarnings('ignore')
```
