### Area Plot
- Area plots, also known as area charts, are primarily used in statistics to visualize trends, compare magnitudes, and show part-to-whole relationships over time or across categories.
- They are particularly effective when displaying cumulative data, proportions, or when emphasizing the magnitude of changes in related data series

### Area Chart Using Matplotlib

```python
import matplotlib.pyplot as plt

year = [2015,2016,2017,2018,2019,2020,2021,2022,2023]
sales = [22000,23000,12000,43000,23900,34300,22000,12234,50000]
sales2 = [24000,29000,20000,29000,20900,30300,25000,19234,45000]

plt.xlabel("Day")
plt.ylabel("Pushup Count")
plt.title("Pushups count by day")
plt.fill_between(year,sales,color='r',linewidth=1,linestyle='dashed')
plt.show()
```
