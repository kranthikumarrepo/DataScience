### Histogram 
- Histograms are powerful graphical tools in statistics used to represent the distribution of numerical data.
- They visually display the frequency of data points <b>within specified intervals or bins</b>, helping to understand the data's shape, central tendency, dispersion, and potential outliers.
- Number of values for a given column(On Y Axis) in a specified range(X Axis).

### Kernel Density Estimation
- Kernel density estimation (KDE) is a statistical method used to estimate the probability density function of a random variable.
- Uses
- 1. Visualizing Data Distributions:
  2. Identifying Outliers:
  3. Generating Synthetic Data:
- .
- <b>Note:</b>
- Probablity density = Continues data = Range
- Probablity = Descrete data = exact point

### Histogram vs KDE
- Histograms and Kernel Density Estimation (KDE) plots are both used to visualize data distributions, but they differ in how they represent the data.
- Histograms use bins to group data points, while KDEs use a kernel function to create a smooth, continuous curve that estimates the probability density of the data.
- KDEs are particularly useful for visualizing continuous data and identifying subtle features like skewness or multimodality, which might be missed by histograms.

```python
# Import tips dataset 

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tips_data = sns.load_dataset('tips')
```

```python
# If we draw a normal bar plot for total bill, it shows each line for each total_bill value, which becomes difficult to understand.

sns.barplot(x=tips_data.total_bill, y=tips_data.tip, data=tips_data)
```

#### Create a Histogram Plot Using Matplotlib

```python
# To see total_bill value for each interval of total_bill instead of each value of total bill, use histogram

plt.hist(tips_data.total_bill, bins=10)
```

#### Create a Histogram Plot using Seaborn

```python
# To see total_bill value for each interval of total_bill instead of each value of total bill, use histogram

sns.histplot(x=tips_data.total_bill,data=tips_data,bins=10)
```

#### Create a KDE Plot using Matplotlinb
- Note:- A histogram can be converted into a distribution chart (probability density chart) just by smoothly connecting the peak of each histogram bar.

```python
# KDE is useful to understand the skewness of the data
# Check how total_bill is distributed.

tips_data.total_bill.skew
```

#### Create a KDE Plot using Seaborn

```python
# Create a histogram for total_bill column along with skewness (KDE) for the same column
# Using histplot to create KDE, just use kde = True parameter.

sns.histplot(x=tips_data.total_bill,data=tips_data, bins = 10, kde=True)
```

```python
# Create a histogram for total_bill column along with skewness (KDE) for the same column
# by directly using kdeplot, this function is not available in matplotlib

sns.kdeplot(x=tips_data.total_bill, levels=15,data=tips_data)
```
