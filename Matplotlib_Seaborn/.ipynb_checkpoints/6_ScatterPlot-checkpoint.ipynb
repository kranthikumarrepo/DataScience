### Scatter Plot
- <b>Shows the relationship between two numerical variable (CoRelation/ Patterns) and also shows outliers.</b>
- Scatter plots are used in statistics to visually represent the relationship between two numerical variables.
- They help to identify patterns, trends, and correlations within the data, and can be used to determine if there is a relationship between the variables and the nature of that relationship (positive, negative, linear, non-linear, strong, or weak)

#### Scatter Plot Uses:-
Displays the relationship between two sets of numerical data. 
Data Representation: Each point on the plot represents a pair of values (x, y). 
Use Cases: Examining correlations between variables (e.g., height and weight), identifying outliers, and exploring data patterns. Example: Plotting the relationship between temperature and ice cream sales.

### Line Charts
- <b>Shows the trend of a variable over time, useful to understand the trend over the time. Useful for trends and prediction </b>
- Line charts are primarily used in statistics to visualize and analyze trends and patterns in data, especially over time.
- They are particularly effective for displaying how a variable changes in relation to another continuous variable, such as time.
- Line charts excel at showing trends, fluctuations, and comparisons between different datasets.

#### Line Chart Uses:-
Shows the trend of a single variable over a continuous interval (e.g., time). 
Data Representation: Points are plotted and connected by lines, often showing how a value changes over time. 
Use Cases: Tracking trends, identifying patterns, and making predictions based on historical data. Example: Tracking the stock price of a company over a year.

#### Line vs Scatter Chart
- Scatter plots and line charts are both used to visualize data, but they serve different purposes.
- Scatter plots show the relationship between two variables, while line charts track changes in a single variable over time or another continuous scale

### Key Differences:
Focus:- Scatter plots emphasize the relationship between two variables, while line charts focus on the trend of a single variable.
Data Type:- Scatter plots can handle any two sets of numerical data, while line charts are typically used for data collected over a continuous interval, such as time.
Interpretation:- Scatter plots are used to identify correlations and patterns, while line charts are used to observe trends and make predictions. 

### Scattered Plot using Matplotlib

```python
# Simple Scattered Plot 

import matplotlib.pyplot as plt

days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
numbers = [99,75,120,150,66,25,0]

plt.xlabel("Day")
plt.ylabel("Pushup Count")
plt.title("Pushups count by day")
plt.scatter(days,numbers,color='r')
```

### Scattered Plot using Seaborn

```python
# Create a scattered plot using the below data

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df_excel = pd.read_excel(r"C:\Users\ACER\Desktop\Kranthi\DataScience_Desktop\Python\Pandas_Files\InputFiles\data_excel_4.xlsx")
df_excel.head(8)
```

```python
# Create a scatter plot based on ID on X axis and Value on Y axis

sns.scatterplot(x='ID', y='Value', data=df_excel)
plt.show()
```

```python
# Create a scatter plot based on ID on X axis and Value on Y axis, differentiate the scattered plots based on category

sns.scatterplot(x='ID', y='Value', hue='Category', data=df_excel, size=99)
plt.show()
```

### Scatter Plot with another dataset using Seaborn

```python
# Import tips dataset 

import seaborn as sns
import pandas as pd

tips_data = sns.load_dataset('tips')
```

```python
tips_data.head(5)
```

```python
# First draw a line plot using the above data set on Total bill on x axis and Tip on y axis

sns.lineplot(x='total_bill', y='tip', hue = 'sex', data=tips_data)
```

```python
# Convert the above line to scattered plot

sns.scatterplot(x='total_bill', y='tip', hue = 'sex', data=tips_data)
```

```python
# Scatter Plot with Line Plot

sns.scatterplot(x='total_bill', y='tip', hue = 'sex', data=tips_data)
sns.lineplot(x='total_bill', y='tip', hue = 'sex', data=tips_data)
```
