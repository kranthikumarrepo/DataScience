### Strip Plot
- Creates a seperate plot for each item (Category) on X axis.
- A strip plot is a single-axis scatter plot that is used to visualise the distribution of many individual one-dimensional values.
- The values are plotted as dots along one unique axis, and the dots with the same value can overlap.
- Numerical Column on one axis(y) and the categorical column on the other axis(x)

```python
# Import tips dataset 

import seaborn as sns
import pandas as pd

tips_data = sns.load_dataset('tips')
```

```python
tips_data.head(5)
```

#### Create a Strip Plot using Matplotlib

- A stripplot, while not a native function in Matplotlib, is a type of scatter plot that visualizes the distribution of a numeric variable by a categorical variable. It is a core feature of the Seaborn library, which is built on top of Matplotlib and provides a higher-level interface for creating statistical graphics.

#### Create a Strip Plot using Seaborn

```python
# Create a strip plot with Time(lunch and dinner) wise total bill

sns.stripplot(x=tips_data.time,y=tips_data.total_bill,hue=tips_data.sex,data = tips_data)
```

```python
# Create a Strip Plot with Sex on x axis and Total bill on Y axis

sns.stripplot(x=tips_data.sex,y=tips_data.total_bill)
```

```python
# Create a Strip Plot with Sex on x axis and Total bill on Y axis, colour by time

sns.stripplot(x=tips_data.sex, y=tips_data.total_bill, hue=tips_data.time, data = tips_data)
```

```python
# Create a Strip Plot - Day wise total bill, colour by sex

sns.stripplot(x=tips_data.day,y=tips_data.total_bill,hue=tips_data.sex, data = tips_data)
```
