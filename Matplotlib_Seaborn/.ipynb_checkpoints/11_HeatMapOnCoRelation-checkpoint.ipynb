### HeatMap Plot
- To see the CoRelation between the columns.
- In statistics, a heatmap is a data visualization technique that uses colors to represent the magnitude of values in a table or matrix.
- It's a way to quickly identify patterns, correlations, and variations within a dataset.
- A correlation heatmap is a visual representation of a correlation matrix, using colors to represent the strength and direction of relationships between variables.
- .
- Note:-
- A correlation matrix is a table showing correlation coefficients between variables, while a heatmap is a visual representation of that matrix, using color to encode the magnitude of the correlations. Essentially, a heatmap is a way to visualize the correlation matrix, making it easier to identify patterns and relationships between variables.
- .
- In Python, the corr() function, most commonly found as a method of a Pandas DataFrame, calculates the pairwise correlation between all the numeric columns of the DataFrame, whereas corrcoef() is used to find the CoRelation between any two columns only.
- .
- Note:-
- HeatMap can be generated using Matplotlib, but it is lengthy, seaborn is very simple.

```python
# Import the required dataset

import seaborn as sns
import numpy as np

taxis_data = sns.load_dataset('taxis')
taxis_data.head(3)
```

#### Calculate the CoRelation between two columns and draw heat map between those two columns

```python
# To calculate the CoRelation between any two columns, use numpy.corrcoef(col1,col2)

CoR_Distance_Tip = np.corrcoef(taxis_data.distance,taxis_data.tip)
print("CoRelation Coeficient Matrix Between Distance and Tip \n", CoR_Distance_Tip)

CoR_Distance_Total = np.corrcoef(taxis_data.distance,taxis_data.total)
print("CoRelation Coeficient Matrix Between Distance and Total Amount \n", CoR_Distance_Total)
```

```python
# To see the above CoRelation in a graphical format, use heat map
# Send the CoRelation Coeficient matrix to the heatmap function.

sns.heatmap(CoR_Distance_Tip,annot=True)
```

```python
# To see the above CoRelation in a graphical format, use heat map
# Send the CoRelation Coeficient matrix to the heatmap function.

sns.heatmap(CoR_Distance_Total,annot=True)
```

#### Calculate the CoRelation between all columns and draw heat map between those columns

```python
# To Calculate the CoRelation between all the numerical columns in a dataframe use corr function

CoR_All_Cols_taxis_data_numeric = taxis_data.corr(numeric_only=True)
print("CoRelation Matrix between all the columns \n\n", CoR_All_Cols_taxis_data_numeric)
```

```python
# To see the above pairwise CoRelation in a graphical format, use heat map
# Send the CoRelation matrix between all the columns to the heatmap function.

sns.heatmap(CoR_All_Cols_taxis_data_numeric,annot=True)
```

```python
# One more example for CoRelation and Heat Map
# Calculate the CoRelation between the columns in below dataframe and then draw the heat map for the calculated coRelation

import pandas as pd
import seaborn as sns
import numpy as np

data = {'A': [1, 2, 3, 4, 5],
        'B': [2, 4, 5, 4, 6],
        'C': [5, 4, 3, 2, 1]}
df = pd.DataFrame(data)
print("Data is \n", df)

# Calculate the CoRelation between the columns
correlation_matrix = df.corr()
print("\nCoRelation Matrix is \n",correlation_matrix)

# Draw the heat map using the calculated corelation matrix
sns.heatmap(correlation_matrix,annot=True)
```
