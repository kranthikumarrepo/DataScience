### Bar/Column Charts
- Bar charts are versatile statistical tools used to visually represent and compare data across different categories or over time.
- They are particularly useful for displaying discrete data, showing frequency distributions, and highlighting differences between groups.

### Column Chart using matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
plt.bar(Company,Revenue)
```

### Column Chart using seaborn

```python
import seaborn as sns

Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]

sns.barplot(x = Company, y = Revenue)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")
df.head()
sns.barplot(x = 'who', y = 'fare',data = df)
plt.show()
```

### Bar Chart(Horizontal) using matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
plt.barh(Company,Revenue)
```

### Bar Chart(Horizontal) using seaborn

```python
import seaborn as sns

Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
sns.barplot(y = Company,x = Revenue, orient = 'h')
```

### Stacked Column Chart, with Label using matplotlib

```python
import matplotlib.pyplot as plt

Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
Profit = [30,40,28,12]
plt.bar(Company,Revenue, label = "Revenue")
plt.bar(Company,Profit, label = "Profit")
plt.legend()
```

### Stacked Column Chart, with Label using Seaborn

```python
import seaborn as sns

Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
Profit =  [10,16,9,7]

sns.barplot(x= Company,y = Revenue,orient='v')
sns.barplot(x = Company,y = Profit, orient='v')
```

### Stacked Bar Chart using matplotlib

```python
import matplotlib.pyplot as plt

Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
Profit = [30,40,28,12]
plt.barh(Company,Revenue, label = "Revenue")
plt.barh(Company,Profit, label = "Profit")
plt.legend()
```

### Stacked Bar Chart using Seaborn

```python
import seaborn as sns

Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
Profit = [30,40,28,12]

sns.barplot(y= Company,x= Revenue,orient='h')
sns.barplot(y=Company,x= Profit,orient='h')
```
