### Column chart using matplotlib
- First create a normal stacked column chart using matplotlib
- Two ways to create
- 1. by directly using the given column names
  2. by converting the categorical data to numbers and then back to categorical - Helpful while creating a clustered charts

```python
# Directly use the column names to create a stacked column chart
# This method is not helpful to create a clustered chart

import matplotlib.pyplot as plt
import numpy as np
Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
Profit = [30,40,28,12]
plt.xlabel("Company")
plt.ylabel("Profit/  Revenue")
plt.bar(Company,Revenue)
plt.bar(Company,Profit)
```

```python
# Create clustered column chart by converting the categorical column names to numbers.

import matplotlib.pyplot as plt
import numpy as np
Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
Profit = [30,40,28,12]

# Convert Company name in text to numbers
Company2Nums = np.arange(len(Company))
plt.title("US Tech Performance")
plt.xlabel("Company")
plt.ylabel("Profit/  Revenue")
#plt.xticks(Company2Nums,Company)
plt.bar(Company2Nums,Revenue)
plt.bar(Company2Nums,Profit)
print(Company2Nums)
```

```python
# Create stacked column chart by converting the categorical column names to numbers.
# Convert back the numbers to categorical column name
# This helps in next steps to create a clustered charts

import matplotlib.pyplot as plt
import numpy as np
Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
Profit = [30,40,28,12]

# Convert Company name in text to numbers
Company2Nums = np.arange(len(Company))

plt.title("US Tech Performance")
plt.xlabel("Company")
plt.ylabel("Profit/  Revenue")

# To convert back the numbers to Categorical column
plt.xticks(Company2Nums,Company)

plt.bar(Company2Nums,Revenue)
plt.bar(Company2Nums,Profit)
#print(Company2Nums)
```

### Clustered column chart using matplotlib

```python
# Create clustered column chart by converting the categorical column names to numbers.
# Convert back the numbers to categorical column name

import matplotlib.pyplot as plt
import numpy as np
Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
Profit = [30,40,28,12]

# Convert Company name in text to numbers
Company2Nums = np.arange(len(Company))

plt.title("US Tech Performance")
plt.xlabel("Company")
plt.ylabel("Profit/Revenue")

# To convert back the numbers to Categorical column
plt.xticks(Company2Nums,Company)

# Create clustered chart by changing the position of Revenue and Profit on x axis
plt.bar(Company2Nums-0.4,Revenue,width=0.4,label="Revenue")
plt.bar(Company2Nums,Profit,width=0.4,label="Profit")
plt.legend()
```

#### Point Estimate

- Exact point estimation
- Point estimate is a single value calculated from sample data that is used to estimate an unknown population parameter.
- It's essentially the "best guess" for the true population value based on the information available from the sample.
- .
- Ex:-
- Exactly Sachin with score 99 in todays match'.
- .
- Probablity = descrete >>> exact point

#### Confidence Interval

- Approx. point estimation.
- A confidence interval in statistics is a range of values calculated from sample data that is likely to contain the true value of an unknown population parameter, like the mean or proportion.
- It provides a measure of uncertainty around an estimate, indicating a range within which the true value is expected to fall with a certain degree of confidence
- A confidence interval is the mean of your estimate plus and minus the variation in that estimate. This is the range of values you expect your estimate to fall between if you redo your test, within a certain level of confidence.
- Confidence, in statistics, is another way to describe probability.
- .
- Ex:-
- 95% confident that todays score will be in range of 25 to 35, this is confidance interval.
- .
- Probablity density = continues data >>> range

### Clustered Bar Chart Using matplotlib

```python
# Create clustered column chart by converting the categorical column names to numbers.
# Convert back the numbers to categorical column name

import matplotlib.pyplot as plt
import numpy as np
Company = ['Google','Amazon','Facebook','Microsoft']
Revenue = [90,136,89,27]
Profit = [30,40,28,12]

# Convert Company name in text to numbers
Company2Nums = np.arange(len(Company))

plt.title("US Tech Performance")
plt.ylabel("Company")
plt.xlabel("Profit/  Revenue")

# To convert back the numbers to Categorical column
plt.yticks(Company2Nums,Company)

# Create clustered chart by changing the position of Revenue and Profit on x axis
plt.barh(Company2Nums-0.4,Revenue,height=0.4,label="Revenue")
plt.barh(Company2Nums,Profit,height=0.4,label="Profit")
plt.legend()
```
