### Random Data, Graphs(Histogram and Distribution plot), Skewness, Kurtosis

```python
# Generate a set of random numbers
import numpy as np
random_data = np.random.normal(25000,15000,1000) # Mean, SD and No. of data points
#print(random_data)
```

```python
# Create a histogram from the above data
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(random_data) # Creates the histogram graph
plt.show() # Shows the graph
```

```python
# Create both histogram and distribution plot using kernal density estimate curve.
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
sns.distplot(random_data)
plt.show()
```

```python
# Check the skewness and kurtosis
from scipy import stats

print("Skewness = ",stats.skew(random_data))
print("Kurtosis = ",stats.kurtosis(random_data))
```

#### From above data, it looks like a random distribution, add huge outliers and make it skewed.

```python
# Calculate Mean and Median for the above normally distributed data
import numpy
print("Mean is ", numpy.mean(random_data))
print("Median is ", numpy.median(random_data))

# Add outliers data to the above random data
random_data = numpy.append(random_data,[9999,999999])
# Calculate Mean and Median after adding outlier data
print("Mean is ", numpy.mean(random_data))
print("Median is ", numpy.median(random_data))

# Add huge outlier data to the above random data
random_data = numpy.append(random_data,[999999999])
# Calculate Mean and Median after adding outlier data
print("Mean is ", numpy.mean(random_data))
print("Median is ", numpy.median(random_data))
```

```python
# Check the skewness and kurtosis
from scipy import stats

print("Skewness = ",stats.skew(random_data))
print("Kurtosis = ",stats.kurtosis(random_data))
```

```python
# After adding huge outliers, create both histogram and distribution plot using kernal density estimate curve.

import seaborn as sns
sns.distplot(random_data)
# sns.distplot(random_data,kde=True) >>> both are same.
plt.show()
```
